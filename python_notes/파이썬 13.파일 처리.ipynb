# 파일 처리

## 파일 열기

파일을 다루기 위해선 우선 파일을 열어야 합니다. Python에서는 `open()` 함수를 사용하여 파일을 열 수 있습니다. 파일을 열기 위한 문법은 다음과 같습니다.
```python
file = open("filename", "mode")
```
여기서, `filename`는 열고자 하는 파일의 이름이며, `mode`는 파일을 열어서 할 짓을 지정합니다. 예를 들어, 파일을 읽기, 쓰기 또는 추가하기 위해 파일을 열 수 있습니다.

파일 열기 모드:

* `r`: 읽기 모드. 기본 모드. 파일을 읽습니다.
* `w`: 쓰기 모드. 파일을 쓰기 위해 엽니다. 파일이 이미 존재하면 있던 내용을 갖다버리고 덮어씁니다. 파일이 없으면 새 파일을 생성합니다.
* `a`: 추가 모드. 파일을 이어서 씁니다. 파일이 없으면 새 파일을 생성합니다.
* `x`: 배타적 생성 모드. 파일을 쓰기 위해 엽니다. 파일이 이미 존재하면 실패합니다.
* `b`: 바이너리 모드. 이진 파일(이미지, 오디오 등)에 사용되는 바이너리 모드로 파일을 엽니다.
* `t`: 텍스트 모드. 텍스트 파일에 사용되는 텍스트 모드로 파일을 엽니다.

예시로, "sample.txt"를 열어 봅시다.

````python
file = open("sample.txt", "w")
````

파일을 열었기 때문에, 옆에 아이콘들에서 파일 탭에 들어가면 sample.txt가 있는 것을 확인할 수 있습니다.

## 파일 닫기

작업을 마친 후에는 연 파일을 닫아야 합니다. Python은 파일을 닫기 위한 내장 함수 `close()`를 가지고 있습니다. 파일을 사용한 후에는 파일을 닫는 것이 좋습니다. 파일을 닫기 위한 문법은 다음과 같습니다:
```python
file.close()
```
넹. 간단합니다.

예시:

````python
file = open("sample.txt", "r")

# 대충 파일가지고 작업하는 코드

file.close()
````

이렇게 해서 파일을 열고 닫는 것에 대해 알아봤습니다.

## 파일 읽기

파일을 열었으면? 파일을 읽어야죠. Python은 파일을 읽기 위해 다양한 메서드를 제공합니다. 가장 일반적인 두 가지 메서드는 다음과 같습니다:

* `read()`: 파일 전체를 읽고, 문자열로 반환합니다.
* `readline()`: 파일에서 한 줄(개행 문자 `\n` 기준)을 읽고, 문자열로 반환합니다.


## 파일 쓰기

읽기뿐만 아니라 쓰기도 해야죠. 파일에 데이터를 쓰기 위해서는 `write()` 메서드를 사용할 수 있습니다. 파일에 쓰기 위한 구문은 다음과 같습니다:
```python
file.write("대충 파일에다가 쓸 데이터들")
```

파일 읽기와 쓰기에 대한 예제입니다. "sample.txt"라는 파일을 생성해 거기에다가 문자열을 쓰고, 읽어보겠습니다.

````python
file = open("sample.txt", "w")
file.write("Hello, World!")
file.close()
````

이 코드를 실행 후 왼쪽 파일에서 sample.txt를 열어보면, Hello, World!가 적힌 것을 확인할 수 있습니다.

````python
file = open("sample.txt", "w")
string = """\
Hello, World!
This is from GSA!
Run away from GSA!!\
"""
file.write(string)
file.close()
````

이 코드를 실행하고 나면 여러 줄의 텍스트가 적힌 것을 알 수 있습니다. 이제 이걸 읽어서 가져와보겠습니다.

`read()` 예시:

````python
file = open("sample.txt", "r")
contents = file.read()
print(contents)
file.close()
````

`readline()` 예시:

````python
file = open("sample.txt", "r")
line = file.readline()
print(line)
file.close()
````

반복문을 이용해 한 줄씩 뗄 수도 있습니다.

````python
file = open("sample.txt", "r")
for line in file:
    print(line)
file.close()
````

## 데이터 이어쓰기

파일을 열 때, `a`를 이용해 덮어쓰는 것이 아닌, 기존에 있는 내용에 데이터를 이어붙일 수 있습니다.

위의 코드들로부터 이어지는 예시입니다:

````python
file = open("sample.txt", "a")
file.write("\nSA RANG HAE YO")
file.close()
````

읽어볼까요

````python
file = open("sample.txt", "r")
contents = file.read()
print(contents)
file.close()
````

잘 이어써진걸 알 수 있습니다.

## `with`문

Python의 `with` 문은 코드 블록이 실행되는 동안 리소스가 사용되고 있음을 보장하고, 예외가 발생하더라도 블록이 종료될 때 적절하게 닫히고 정리되도록 하는 방법입니다. 이는 일반적으로 파일과 함께 사용되며, 파일이 올바르게 열리고 닫히는 것을 보장합니다.

파일과 함께 `with`를 사용하는 일반적인 구문은 다음과 같습니다.
```python
with open(file_path, mode) as file:
    # 파일가지고 할거하는 코드
```
여기서 `file_path`는 열고자 하는 파일의 경로이며, `mode`는 파일을 읽을 것인지 쓸 것인지를 지정하고, `file`은 작업할 파일 객체입니다.

이렇게 `with`를 사용하면 블록이 종료될 때 예외가 발생하더라도 파일이 올바르게 닫히는 것을 보장합니다. 즉, `close()`안해도 자동으로 닫아줍니다!

위의 "sample.txt"를 읽는 예시입니다:

````python
with open("sample.txt", "r") as file:
    contents = file.read()
    print(contents)
````

````python
!gdown https://drive.google.com/uc?id=1HaQtiJ2f1P3DyZXRm7WeHQ8r515mQCVj
# ↑ input.txt 파일 다운로드하는거
# 무시하세용

# 입출력 파일 열기
# 이것과 같이 두 개 이상을 한번에 할 수 있습니다 
with open('input.txt', 'r') as input_file, open('output.txt', 'w') as output_file:
    # 파일의 줄마다 반복
    for line in input_file:
        # 알파벳 아닌거 지우고 소문자로
        clean_line = ''.join(filter(str.isalpha, line)).lower()
        # 처리한 문자들 쓰기
        output_file.write(clean_line + '\n')

with open('output.txt', 'r') as file:
    contents = file.read()
    print(contents)
````
