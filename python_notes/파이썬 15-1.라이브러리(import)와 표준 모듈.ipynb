# 모듈

모듈은 Python의 definition과 statement들을을 담고 있는 파일입니다. 파일 이름은 확장자 `.py`가 추가된 모듈 이름입니다. 예를 들어, `mymodule.py`라는 모듈이 있다면 `import` 문을 사용하여 다른 파이썬 파일에서 가져올 수 있습니다.

## 모듈 임포트

`import` 문을 사용하여 파이썬 스크립트에 모듈을 가져올 수 있습니다. 모듈을 가져오는 구문은 다음과 같습니다:
```python
import module_name
```
여기서 `module_name`은 가져오려는 모듈의 이름입니다. 모듈을 가져온 후, 점(.) 표기법을 사용하여 해당 모듈의 함수와 변수에 액세스할 수 있습니다. 예를 들어, `mymodule`이라는 모듈에 `my_function`이라는 함수가 있다면, 다음과 같이 액세스할 수 있습니다:
```python
import mymodule

mymodule.my_function()
```

## 모듈에서 특정 함수를 갖고오기

다음 구문을 사용하여 모듈에서 특정 함수를 가져올 수도 있습니다:
```python
from module_name import function_name
```
여기서 `module_name`은 가져올 모듈의 이름이며, `function_name`은 가져올 함수의 이름입니다. 함수를 가져온 후, 모듈 이름을 사용하지 않고 직접 함수를 사용할 수 있습니다. 예를 들어, `mymodule`이라는 모듈에 `my_function`이라는 함수가 있다면, 다음과 같이 가져와 사용할 수 있습니다:
```python
from mymodule import my_function

my_function()
```

## 모듈 또는 함수 이름 변경하기

`as` 키워드를 사용하여 모듈 또는 함수 이름을 가져올 때 이름을 변경할 수도 있습니다. 모듈 또는 함수 이름이 너무 길거나 다른 모듈 또는 함수와 이름 충돌이 발생하는 경우 유용합니다. 다음은 예시입니다:
```python
import mymodule as mm

mm.my_function()
```
이 예시에서는 `mymodule`을 `mm`으로 이름을 변경하였습니다. 이제 `mymodule` 대신 `mm`을 사용하여 해당 모듈의 함수에 액세스할 수 있습니다.

마찬가지로, 함수를 가져올 때 다음과 같이 이름을 변경할 수 있습니다:
```python
from mymodule import my_function as mf

mf()
```
여기서는, `my_function`을 `mf`로 변경하였습니다.

이 기능을 다음과 같이 많이 사용합니다 ~얘내는 국룰임~:
```python
import numpy as np
import matplotlib.pyplot as plt
```

## 모듈의 모든 함수를 갖고오기

만약 모듈에서 모든 함수를 가져오고 싶다면, 다음과 같은 형식을 사용할 수 있습니다.
```python
from module_name import *
```
여기서 별표`*` 기호는 "모두 가져오기"를 의미합니다. 하지만 이 구문을 사용하는 것은 이름 충돌이 발생할 수 있으며 코드를 읽고 이해하기 어렵게 만들 수 있기 때문에 일반적으로 권장되지 않습니다. 대신, 이전에 논의한 구문을 사용하여 필요한 함수만 가져오는 것이 좋습니다.

# `random` 모듈
`random` 라이브러리는 파이썬 내장 모듈로, 무작위 숫자, 무작위 선택, 무작위 시퀀스를 생성하는 다양한 함수 모음을 제공합니다. 이 모듈은 시뮬레이션, 암호학, 게임 등 다양한 응용 분야에서 널리 사용됩니다.

## `random` 모듈 가져오기
Python 스크립트에서 `random` 라이브러리를 사용하려면, `import` 문을 사용하여 라이브러리를 가져와야 합니다.
```python
import random
```
`random` 라이브러리를 가져왔으면, 이제 해당 라이브러리의 함수와 메소드를 사용하여 무작위 수 및 시퀀스를 생성할 수 있습니다.

## 랜덤 숫자 생성
`random` 라이브러리는 다양한 유형의 무작위 숫자를 생성하는 몇 가지 함수를 제공합니다. 다음은 가장 일반적으로 사용되는 일부 함수입니다:

### `random()`
`random()` 함수는 0부터 1까지(0은 포함되지만 1은 포함되지 않는) 무작위 실수를 반환합니다:

````python
import random

x = random.random()
print(x)
````

### `randint()`
`randint(a, b)` 함수는 `a`부터 `b`까지(둘 다 포함) 무작위 정수를 반환합니다:

````python
import random

x = random.randint(1, 10)
print(x)
````

### `randrange()`
`randrange(start, stop, step)` 함수는 `start`, `stop`, `step` 인수로 생성된 범위에서 무작위로 선택된 요소를 반환합니다:

````python
import random

x = random.randrange(0, 101, 10)
print(x)
````

### `uniform()`
`uniform(a, b)` 함수는 `a`부터 `b`까지(`a`는 포함되지만 `b`는 포함되지 않는) 무작위 실수를 반환합니다:

````python
import random

x = random.uniform(0, 5)
print(x)
````

## 랜덤 시퀀스 생성
`random` 라이브러리는 다양한 유형의 무작위 시퀀스를 생성하는 몇 가지 함수를 제공합니다. 다음은 가장 일반적으로 사용되는 일부 함수입니다:

### `choice()`
`choice(seq)` 함수는 비어 있지 않은 시퀀스 seq에서 무작위 요소를 반환합니다:

````python
import random

my_list = [1, 2, 3, 4, 5]
x = random.choice(my_list)
print(x)
````

### `shuffle()`
`shuffle(seq)` 함수는 변경 가능한 시퀀스 `seq`의 요소를 임의로 섞습니다:

````python
import random

my_list = [1, 2, 3, 4, 5]
random.shuffle(my_list)
print(my_list)
````

### `sample()`
`sample(population, k)` 함수는 비어 있지 않은 `population` 시퀀스에서 무작위로 선택된 k개의 고유한 요소를 포함하는 새 리스트를 반환합니다:

````python
import random

my_list = [1, 2, 3, 4, 5]
x = random.sample(my_list, 3)
print(x)
````

## 랜덤 시드 설정
기본적으로 `random` 라이브러리는 무작위 수 생성기의 시드로 현재 시스템 시간을 사용합니다. 그러나 `seed(a=None)` 함수를 사용하여 명시적으로 시드를 설정할 수 있습니다:

````python
import random

random.seed(42)
x = random.random()
print(x)
````

※ 시드란?

컴퓨터 과학에서, 시드는 난수 생성기 또는 다른 결정론적 알고리즘의 입력값으로 사용되어 보이는 난수의 수열을 생성하는 초기값입니다. 같은 시드는 항상 같은 수열을 생성하므로 재현 가능한 결과를 얻을 수 있습니다.

# os 모듈

Python의 os 모듈은 운영 체제와 상호 작용하기 위한 다양한 기능을 제공합니다. 파일 및 디렉터리 관리, 프로세스 관리, 환경 변수 등과 관련된 다양한 작업을 수행할 수 있습니다.

## os 모듈 가져오기
os 모듈을 사용하기 전에 Python 스크립트로 가져와야 합니다. 다음 코드로 os 모듈을 가져올 수 있습니다.

```python
import os
```

## 파일 및 디렉토리 작업

os 모듈은 파일과 디렉토리 작업을 위한 함수를 제공합니다. 일반적으로 사용되는 일부 기능은 다음과 같습니다.

~시험엔 거의 안나와요~

* **`os.name`**: 현재 운영체제
* **`os.getcwd()`**: 현재 작업 디렉토리를 문자열로 반환합니다.
* `os.chdir(경로`): 현재 작업 디렉토리를 지정된 경로로 변경합니다.
* `os.listdir(경로)`: 지정된 경로에 있는 파일 및 디렉토리 목록을 반환합니다.
* **`os.mkdir(경로)`**: 지정된 경로로 디렉토리를 생성합니다.
* `os.makedirs(경로)`: 디렉토리를 재귀적으로 생성하여 필요한 모든 중간 디렉토리를 생성합니다.
* **`os.remove(경로)`**: 지정된 경로에서 파일을 제거합니다.
* **`os.rmdir(경로)`**: 지정된 경로에서 빈 디렉토리를 제거합니다.
* `os.removedirs(경로)`: 디렉터리를 재귀적으로 제거하여 빈 중간 디렉터리를 모두 제거합니다.
* **`os.rename(원천, 대상)`**: 원천 경로에서 대상 경로로 파일 또는 디렉토리의 이름을 바꿉니다.
* `os.path.exists(경로)`: 지정된 경로가 존재하면 True를, 그렇지 않으면 False를 반환합니다.
* `os.path.isfile(경로)`: 지정된 경로가 파일이면 True를 반환하고 그렇지 않으면 False를 반환합니다.
* `os.path.isdir(경로)`: 지정된 경로가 디렉토리이면 True를 반환하고 그렇지 않으면 False를 반환합니다.

````python
import os

# 간단한 정보 출력
print("현재 운영체제:", os.name)
print("현재 폴더:", os.getcwd())
print("현재 폴더 내부의 요소:", os.listdir())

# 디렉토리(폴더) 만들고 지우기
os.mkdir("hello")
print("현재 폴더 내부의 요소:", os.listdir())
os.rmdir("hello")
print("현재 폴더 내부의 요소:", os.listdir())

# 파일 만들고 이름바꿨다가 지우기
with open("original.txt", "w") as file:
    file.write("hello")
print("현재 폴더 내부의 요소:", os.listdir())
os.rename("original.txt", "new.txt")
print("현재 폴더 내부의 요소:", os.listdir())
os.remove("new.txt")
print("현재 폴더 내부의 요소:", os.listdir())

# 시스템 명령어 사용
os.system("dir")
# ↑Windows 환경에서는 명령 프롬프트에 dir를 입력했을 때와 같은 결과가 나옴
# Colab 환경이 Windows가 아니라서 0나옵니다
````

````python
import os

with open("original.txt", "w") as file:
    file.write("hello")

output = os.listdir(".")

print("os.listdir():", output)
print()

print("# 폴더와 파일 구분하기")
for path in output:
    if os.path.isdir(path):
        print("폴더:", path)
    else:
        print("파일:", path)
````

````python
import os

# 디렉토리 내의 모든 파일 출력하기

def read_folder(path):
    # 폴더 요소 읽어들이기
    output = os.listdir(path)
    # 폴더 요소 구분하기
    for item in output:
        # 디렉토리(폴더)면 계속 들어가기
        if os.path.isdir(item):
            read_folder(item)
        # 파일 출력
        else:
            print("파일:", item)


read_folder(".")
````

# datetime 모듈
주로 현재 시간을 불러올 때 사용됩니다.

````python
from datetime import datetime

now = datetime.now()
print("# 시간을 포맷에 맞춰 출력하기")

output_a = now.strftime("%Y.%m.%d %H:%M:%S")

output_b = "{}년 {}월 {}일 {}시 {}분 {}초".format(
    now.year, now.month, now.day, now.hour, now.minute, now.second

)
output_c = now.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초")

print(output_a)
print(output_b)
print(output_c)
````
