## 창 띄우기

Tkinter 응용 프로그램 구축의 첫 번째 단계는 창을 만드는 것입니다. 기본 창을 루트(root) 창이라고 합니다. 간단한 창을 만드는 방법은 다음과 같습니다.

```python
import tkinter as tk

# root 창 만들기
window = tk.Tk()

# 창 이름 설정
window.title("My Window")
# 창 크기 및 위치 설정
window.geometry("400x300+150+100")

# 이벤트 루프 실행
window.mainloop()
```

## 위젯 추가하기

Tkinter는 레이블, 버튼, 텍스트 상자, 확인란 등과 같이 GUI에 추가할 수 있는 다양한 위젯을 제공합니다. Tkinter에서 제공하는 위젯들로는 다음과 같은 것들이 있습니다:

* Button : 간단한 버튼으로 명령을 수행할 때 사용
* Canvas : 화면에 무언가를 그릴 때 사용
* Checkbutton : 2가지의 구별되는 값을 가지는 변수 표현
* Entry : 한 줄의 텍스트를 입력받는 필드
* Frame
  * 컨테이너 클래스
  * 경계선과 배경을 가짐.
  * 다른 위젯들을 그룹핑하는데 사용
* Label : 텍스트나 이미지 표시
* Listbox : 선택사항 표시
* Menu : 메뉴 표시. 풀다운 메뉴나 팝업 메뉴 가능
* Menubutton : 메뉴 버튼. 풀다운 메뉴 가능
* Message : 텍스트 표시. 레이블 위젯과 비슷함. 자동으로 주어진 크기로 텍스트 축소 가능
* Radiobutton : 단일 옵션 선택
* Scale : 슬라이더를 끌어서 수치값을 입력하는데 사용
* Scrollbar : 캔버스, 엔트리, 리스트 박스, 텍스트 위젯을 위한 스크롤바 제공
* Text : 형식을 가지는 텍스트 표시. 여러가지 스타일과 속성으로 텍스트 표시.
* Toplevel : 최상위 윈도우로 표시되는 독립적인 컨테이너 위젯
* LabelFrame : 경계선과 제목을 가지는 프레임 위젯의 변형
* Panedwindow : 자식 위젯들을 크기조절이 가능한 패널로 관리하는 컨테이너 위젯
* Spinbox : 특정한 범위에서 값을 선택하는 엔트리 위젯의 변형

다음은 레이블(설명) 하나와 버튼을 표시하는 간단한 예시입니다.
```python
import tkinter as tk

window = tk.Tk()

# 레이블 만들기
label = tk.Label(window, text="Hello, Tkinter!")
label.pack()

# 버튼 추가하기
button = tk.Button(window, text="Click Me")
button.pack()

window.mainloop()
```

## 레이아웃

Tkinter는 창 내에서 위젯을 구성할 수 있는 여러 레이아웃 관리자를 제공합니다. 가장 일반적으로 사용되는 레이아웃 관리자는 `pack()`, `grid()` 및 `place()`입니다. 각각에 대한 간략한 설명은 다음과 같습니다.

* `pack()`: 위젯을 수평 또는 수직 방식으로 구성하고 차례로 쌓습니다.
* `grid()`: 그리드와 같은 구조로 위젯을 구성하므로 정확한 위젯 배치를 위해 행과 열을 지정할 수 있습니다.
* `place()`: 창 내의 특정 좌표에 위젯을 배치할 수 있습니다.

### `pack()`

`pack()` 레이아웃 관리자는 위젯을 가로 또는 세로 방식으로 구성하고 차례로 쌓습니다. 콘텐츠에 따라 위젯의 크기를 자동으로 조정합니다. 다음은 `pack()` 레이아웃 관리자를 보여주는 예시입니다.

```python
import tkinter as tk

window = tk.Tk()
window.title("pack")
window.geometry("200x100")

label1 = tk.Label(window, text="Label 1")
label1.pack()

button1 = tk.Button(window, text="Button1")
button1.pack()

label2 = tk.Label(window, text="Label 2")
label2.pack()

window.mainloop()
```

이 예에서 위젯들은 `pack()`한 순서대로 세로로 쌓입니다. 또한 `side` 매개변수를 사용하여 위젯을 창의 위쪽, 아래쪽, 왼쪽 또는 오른쪽에 배치할지 여부를 지정할 수 있습니다.

```python
import tkinter as tk

window = tk.Tk()
window.title("pack")
window.geometry("200x100")

label1 = tk.Label(window, text="Label 1")
label1.pack()

button1 = tk.Button(window, text="Button1")
button1.pack(side="right")

label2 = tk.Label(window, text="Label 2")
label2.pack(side="left")

window.mainloop()
```

### `grid()`

`grid()` 레이아웃 관리자는 그리드와 유사한 구조로 위젯을 구성하여 정확한 위젯 배치를 위해 행과 열을 지정할 수 있습니다. `row` 및 `column` 매개변수를 사용하여 그리드 내에서 각 위젯의 위치를 ​​지정할 수 있습니다. 다음은 `grid()` 레이아웃 관리자를 보여주는 예입니다.
```python
import tkinter as tk

window = tk.Tk()
window.title("grid")
window.geometry("200x100")

label1 = tk.Label(window, text="Label 1")
label1.grid(row=0, column=0)

button1 = tk.Button(window, text="Button 1")
button1.grid(row=0, column=1)

label2 = tk.Label(window, text="Label 2")
label2.grid(row=1, column=0, columnspan=2)

window.mainloop()
```

### `place()`

`place()` 레이아웃 관리자를 사용하면 창 내의 특정 좌표에 위젯을 배치할 수 있습니다. `x` 및 `y` 매개변수를 사용하여 각 위젯의 위치를 ​​지정할 수 있습니다. 다음은 `place()` 레이아웃 관리자를 보여주는 예입니다.
```python
import tkinter as tk

window = tk.Tk()
window.title("place")
window.geometry("300x200")

label1 = tk.Label(window, text="Label 1")
label1.place(x=50, y=50)

label2 = tk.Label(window, text="Label 2")
label2.place(x=150, y=100)


window.mainloop()
```

## 이벤트 관리

Tkinter를 사용하면 버튼 클릭, 마우스 이동 및 키 누름과 같은 이벤트에 응답할 수 있습니다. 이벤트 핸들러를 위젯과 연결하여 이벤트가 발생할 때 특정 작업을 수행할 수 있습니다. 다음은 버튼 클릭 이벤트 처리를 보여주는 예입니다.
```python
from tkinter import Tk

from tkinter.ttk import Label, Button, Entry  # tkinter.ttk: Tkinker 확장
# 좀 더 이쁨

def button_click():
    value = entry.get()
    lbl.config(text="입력: " + value)


win = Tk()
win.geometry("300x200")

lbl = Label(win, text="입력: ")
entry = Entry(win)
btn = Button(win, text="확인", command=button_click)

lbl.pack()
entry.pack()
btn.pack()

win.mainloop()
```

## 이벤트 바인딩

창 생성, 위젯 추가 및 레이아웃 관리 외에도 Tkinter를 사용하면 이벤트를 특정 위젯에 바인딩할 수 있습니다. 이를 통해 버튼 클릭, 마우스 이동 및 키보드 누르기와 같은 사용자 상호작용에 반응할 수 있습니다. 다음은 이벤트 바인딩을 보여주는 예입니다.
```python
from tkinter import Tk, Label

x = 150
y = 100

def leftClick(event):
    global x
    x -= 10
    label.place(x=x, y=y)
    print("{}, {}".format(x, y))

def wheelClick(event):
    global x
    x = event.x
    label.place(x=x, y=y)
    print("{}, {}".format(x, y))

def rightClick(event):
    global x
    x += 10
    label.place(x=x, y=y)
    print("{}, {}".format(x, y))

window = Tk()
window.geometry("300x200")

label = Label(window, text="asdf")
label.place(x=x, y=y)

window.bind("<Button-1>", leftClick)
window.bind("<Button-2>", wheelClick)
window.bind("<Button-3>", rightClick)

window.mainloop()
```

## 컨테이너 화면에 이미지 표시

Tkinker의 `PhotoImage()`메서드를 이용해 화면에 이미지를 표시할 수 있습니다. 다음은 이를 보여주는 예시입니다.

<center>

```python
from tkinter import Tk, PhotoImage, Label, Button

def changed(newImg):
    imgLabel.configure(image=newImg, width=600, height=350)
    imgLabel.image = newImg

def img1():
    newImg = PhotoImage(file="cat.png")
    changed(newImg)

def img2():
    newImg = PhotoImage(file="dog.png")
    changed(newImg)

def img3():
    newImg = PhotoImage(file="rubberduck.png")
    changed(newImg)

win = Tk()
win.geometry("600x400")

image = PhotoImage(file="cat.png")
imgLabel = Label(win, image=image, width=600, height=350)

btn_cat = Button(win, text="Cat", command=img1)
btn_dog = Button(win, text="Dog", command=img2)
btn_rubberduck = Button(win, text="Rubber Duck", command=img3)

imgLabel.grid(row=0, column=0, columnspan=3)
btn_cat.grid(row=1, column=0)
btn_dog.grid(row=1, column=1)
btn_rubberduck.grid(row=1, column=2)

win.mainloop()
```
