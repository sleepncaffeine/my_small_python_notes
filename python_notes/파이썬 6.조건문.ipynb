# 조건문

Python에서 조건문은 특정 조건이 참일 경우 코드 블록을 실행하는 데 사용됩니다.<br> 얘내들은 사용자 입력 또는 기타 요인에 따라 결정을 내리는 프로그램을 작성하는 데 필수적입니다.

## if문

if 문은 단일 조건이 참이면 코드 블록을 실행하는 데 사용됩니다.<br>if 문의 기본 구문은 다음과 같습니다.
```python
if 조건:
    # 조건이 참일 시 실행할 코드
```
다음은 예시입니다:

````python
age = 18

if age >= 18:
    print("You are old enough to vote!")
````

위 예시에서는, if 문 안의 코드는 조건 `age >= 18`이 참일 경우에만 실행됩니다.

## if-else문

If-else 문은 조건이 참인 경우 한 블록의 코드를 실행하고, 조건이 거짓인 경우 다른 블록의 코드를 실행하는 데 사용됩니다.<br>if-else 문의 기본 구문은 다음과 같습니다.
```python
if 조건:
    # 조건이 참일 경우 실행할 코드
else:
    # 조건이 거짓일 경우 실행할 코드
```
다음은 예시입니다:

````python
age = 16

if age >= 18:
    print("You are old enough to vote!")
else:
    print("You are not old enough to vote.")
````

이 예제에서, if문 내부의 코드는 `age >= 18` 조건이 참일 때만 실행됩니다.<br>반면 else문 내부의 코드는 `age >= 18` 조건이 거짓일 때 실행됩니다.

`else`는 위에 `if`를 사용했을 때에만 사용할 수 있다는 것에 주의하세요.

## 중첩 if문

중첩 if 문은 다른 if 문 내에 포함된 if 문입니다. 보통 여러 조건에 대해 검사해야 할 때 사용됩니다.<br>중첩 if 문의 기본 구문은 다음과 같습니다.
```python
if 조건1:
    # 조건1이 참일 시 실행할 코드
    if 조건2:
        # 조건1, 조건2 모두가 참일 시 실행할 코드
```
다음은 예시입니다:

````python
age = 18
gender = "female"

if age >= 18:
    if gender == "male":
        print("You are an adult male.")
    else:
        print("You are an adult female.")
else:
    print("You are not yet an adult.")
````

이 예시에서, 중첩 if문 내부의 코드는 조건 `age >= 18` 과 `gender == "male"`이 둘 다 참인 경우에만 실행됩니다.

## 여러 조건문을 가지는 if문

and, or, not과 같은 논리 연산자를 사용하여 if 문에서 여러 조건을 테스트할 수 있습니다.<br>여러 조건을 가진 if 문의 기본 구문은 다음과 같습니다.
```python
if 조건1 and 조건2:
    # 조건1 그리고 조건2가 모두 참일 때 실행할 코드
```
다음은 예시입니다:

````python
age = 18
gender = "female"

if age >= 18 and gender == "female":
    print("You are an adult female.")
else:
    print("You are not an adult female.")
````

이 예시에서, if문 내부의 코드는 조건 `age >= 18` 과 `gender == "female"`이 둘 다 참인 경우에만 실행됩니다.
