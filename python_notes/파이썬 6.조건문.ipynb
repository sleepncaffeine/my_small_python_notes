# 조건문

Python에서 조건문은 특정 조건이 참일 경우 코드 블록을 실행하는 데 사용됩니다.<br> 얘내들은 사용자 입력 또는 기타 요인에 따라 결정을 내리는 프로그램을 작성하는 데 필수적입니다.

## if문

if 문은 단일 조건이 참이면 코드 블록을 실행하는 데 사용됩니다.<br>if 문의 기본 구문은 다음과 같습니다.
```python
if 조건:
    # 조건이 참일 시 실행할 코드
```
다음은 예시입니다:

````python
age = 18

if age >= 18:
    print("You are old enough to vote!")
````

위 예시에서는, if 문 안의 코드는 조건 `age >= 18`이 참일 경우에만 실행됩니다.

## if-else문

If-else 문은 조건이 참인 경우 한 블록의 코드를 실행하고, 조건이 거짓인 경우 다른 블록의 코드를 실행하는 데 사용됩니다.<br>if-else 문의 기본 구문은 다음과 같습니다.
```python
if 조건:
    # 조건이 참일 경우 실행할 코드
else:
    # 조건이 거짓일 경우 실행할 코드
```
다음은 예시입니다:

````python
age = 16

if age >= 18:
    print("You are old enough to vote!")
else:
    print("You are not old enough to vote.")
````

이 예제에서, if문 내부의 코드는 `age >= 18` 조건이 참일 때만 실행됩니다.<br>반면 else문 내부의 코드는 `age >= 18` 조건이 거짓일 때 실행됩니다.

`else`는 위에 `if`를 사용했을 때에만 사용할 수 있다는 것에 주의하세요.

## elif문

파이썬에서 `elif`는 "else if"의 줄임말입니다. `elif`는 `if`와 `else`와 함께 조건문의 일부입니다. 이전 조건이 충족되지 않으면 추가 조건을 지정할 수 있도록 합니다.

`elif`문의 기본 구문은 다음과 같습니다.
```python
if 조건1:
    # 조건1이 참일 시 실행할 코드
elif 조건2:
    # 조건1이 거짓이고 조건2가 참일 시 실행할 코드
else:
    # 조건1, 2 모두 거짓일 시 실행할 코드
```
여기서 `elif` 문은 첫 번째 조건이 거짓인 경우 새로운 조건을 지정하는 데 사용됩니다. 두 번째 조건이 참이면 `elif` 문과 연관된 코드 블록이 실행됩니다. 두 번째 조건이 거짓이면 `else` 블록의 코드가 실행됩니다.

여러 조건을 검사하기 위해 다음과 같이
```python
if 조건1:
elif 조건2:
elif 조건3:
else:
```
하나의 `if` 문에 여러 `elif` 문을 사용할 수 있습니다.

조건문에서 각 조건의 순서는 중요합니다. Python은 순서대로 각 조건을 검사하여 참인 조건을 찾으면 해당 조건 블록의 코드가 실행됩니다. 그리고 이후의 모든 조건들은 무시된다는 것을 기억하십시오.

다음은 예시입니다.

````python
score = 85

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print('Your grade is:', grade)
````

````python
num = float(input("Enter a number: "))

if num > 0:
    print("Positive number")
elif num == 0:
    print("Zero")
else:
    print("Negative number")
````

## 중첩 if문

중첩 if 문은 다른 if 문 내에 포함된 if 문입니다. 보통 여러 조건에 대해 검사해야 할 때 사용됩니다.<br>중첩 if 문의 기본 구문은 다음과 같습니다.
```python
if 조건1:
    # 조건1이 참일 시 실행할 코드
    if 조건2:
        # 조건1, 조건2 모두가 참일 시 실행할 코드
```
다음은 예시입니다:

````python
age = 18
gender = "female"

if age >= 18:
    if gender == "male":
        print("You are an adult male.")
    else:
        print("You are an adult female.")
else:
    print("You are not yet an adult.")
````

이 예시에서, 중첩 if문 내부의 코드는 조건 `age >= 18` 과 `gender == "male"`이 둘 다 참인 경우에만 실행됩니다.

## 여러 조건문을 가지는 if문

and, or, not과 같은 논리 연산자를 사용하여 if 문에서 여러 조건을 테스트할 수 있습니다.<br>여러 조건을 가진 if 문의 기본 구문은 다음과 같습니다.
```python
if 조건1 and 조건2:
    # 조건1 그리고 조건2가 모두 참일 때 실행할 코드
```
다음은 예시입니다:

````python
age = 18
gender = "female"

if age >= 18 and gender == "female":
    print("You are an adult female.")
else:
    print("You are not an adult female.")
````

이 예시에서, if문 내부의 코드는 조건 `age >= 18` 과 `gender == "female"`이 둘 다 참인 경우에만 실행됩니다.
