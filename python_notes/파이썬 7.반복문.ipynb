# 반복문(루프)

Python에서 반복문은 특정 조건이 충족될 때까지 코드 블록을 반복적으로 실행하는 데 사용됩니다.<br>파이썬에는 `while` 반복문과 `for` 반복문 두 가지 종류의 반복문이 있습니다.

~용어 '반복문'과 '루프' 혼용을 좀 하겠습니다~

## `range()` 함수

반복문에 들어가기 전에, 필수적으로 알아야 할 `range()`함수에 대해 알아보겠습니다.

`range()` 함수는 Python의 내장 함수로서, 일련의 숫자를 생성합니다.<br>
이는 최대 다음 3개의 인수를 받을 수 있습니다:
```python
range(start, stop, step)
```
* `start`: (선택) 시퀀스의 시작 숫자입니다. 기본값은 0입니다.
* `stop`: (필수) 시퀀스의 끝 숫자입니다 (시퀀스에 포함되지 않음).
* `step`: (선택) 시퀀스의 각 숫자 간의 차이입니다. 기본값은 1입니다.

\* 문자열에서의 슬라이싱과 비슷하게 생각할 수 있습니다.


다음은 예시입니다:
* `range(1, 5)`는 `1, 2, 3, 4`를 반환합니다. 끝 숫자 `5`를 포함하지 않음을 주의해야 합니다.
* `range(0, 10, 2)`는 `0, 2, 4, 6, 8`을 반환합니다.
* `range(5, 1, -1)`는 `5, 4, 3, 2`를 반환합니다. 이와 같이 `step`을 음수로 설정하면 역순으로 숫자를 생성할 수 있습니다.

※`range()` 함수는 `range` 객체를 반환한다는 점에 유의해야 합니다. 이 객체는 리스트가 아니기 때문에, 리스트로 사용하려면 `list()` 함수를 사용하여 변환해야 합니다. 리스트 자료형에 관해서는 다음에 다루겠습니다.

이와 같은 `range()` 함수는 일련의 숫자를 반복하는 루프에서 자주 사용됩니다.

## `while`반복문

Python에서 `while` 반복문은 반복문의 헤더에서 지정된 조건이 참인 동안 계속 코드 블록을 반복해서 실행합니다.<br>`while` 루프의 기본 구문은 다음과 같습니다:
```python
while 조건:
    # 실행될 코드들
```
`조건`은 불(bool) 자료형 표현식으로 루프 각 반복 전에 검사됩니다.<br> `조건`이 참(`True`)이면 루프 내부의 코드 블록이 실행됩니다. `조건`이 거짓(`False`)이 될 때까지 루프가 계속 실행됩니다.

다음은 1에서 5까지의 숫자를 출력하는 `while` 루프의 예입니다.

````python
i = 1
while i <= 5:
    print(i)
    i += 1
````

위 예시는 매 반복마다 `i`를 출력하고 `i`값에 1을 더합니다. `while`문의 조건이 `i<=5`이기 때문에, `i`가 1, 2, 3, 4, 5 일 때 코드가 실행되고, 그 후 `i`가 6일 때는 실행되지 않습니다.

## `for`반복문

Python의 `for` 루프는 값을 반복하면서 순회하는 데 사용됩니다.<br>`for` 루프의 기본 구문은 다음과 같습니다.
```python
for 변수 in 시퀀스:
    # 실행될 코드들
```
`시퀀스`(sequence)는 리스트, 튜플, 문자열 또는 기타 반복 가능한 객체(iterable object)가 될 수 있습니다.<br>`변수`는 `시퀀스`의 각 값을 하나씩 취하며, 루프 내의 코드들은 각 값마다 실행됩니다.

다음은 숫자 리스트에 대해 반복하고 각 숫자를 출력하는 `for` 반복문의 예시입니다.<br><small>~리스트 아직 쓰기전이긴 한데 이정도는 봐줘요~</small>

````python
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    print(num)
````

위 예시에서 `num`에는 각 실행마다 `numbers`의 값인 1, 2, 3, 4, 5가 들어가 출력됩니다.

다음 예시는 문자열에 대한 `for`반복문입니다.

````python
string = "Hello"
for letter in string:
    print(letter)
````

마찬가지로, 위 예시에서 `letter`에는 각 실행마다 `string`의 원소인 H, e, l, l, o가 순서대로 들어가 출력됩니다.

다음 예시는 위에서 설명한 `range()`함수를 이용한 예시입니다.

````python
for num in range(0, 4):
    print(num, end='')

print('')   # 줄바꿈

for num in range(1, 10, 2):
    print(num, end='')
````

`range()` 함수와 `len()`함수를 이용해 다음과 같은 방법으로도 이용할 수 있습니다.

````python
string = "Hello"
for i in range(len(string)):
    print(string[i], end=',')
````

## 중첩 반복문

중첩 반복문은 말 그대로 반복문 안에 다른 반복문을 넣은 것입니다.

다음은 1부터 3까지의 두 수의 가능한 모든 조합을 출력하는 중첩 반복문의 예시입니다.

````python
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
````

이러한 중첩 반복문의 예시로 국룰인 것이 다음과 같은 별 찍기가 있습니다.<br>
~시험에 잘 나올 만하니까 익혀둡시다~

````python
n = int(input())
for i in range(n):
    for j in range(i+1):
        print('*', end='')
    print('')
````

````python
n = int(input())
for i in range(1, n+1):
    for j in range(n-i):
        print(' ', end="")
    for j in range(i):
        print('*', end="")
    print('')
````

## 반복 제어문

Python은 두 가지 반복문 제어문인 `break`와 `continue`를 제공합니다.

`break` 문은 반복문을 조기 종료하는 데 사용됩니다. 루프 내에서 `break` 문이 실행되면 즉시 루프가 종료되고 프로그램은 반복문 다음 코드로 진행됩니다.

다음은 사용자가 "quit"을 입력할 때 루프를 종료하기 위해 `break` 문을 사용하는 `while` 루프의 예시입니다.

````python
while True:
    response = input("Enter a word (or 'quit' to exit): ")
    if response == 'quit':  # 'quit'이면
        break   # 종료하기
    print("You entered:", response)
````

`continue` 문은 루프의 현재 반복을 건너뛰고 다음 반복으로 계속 진행할 때 사용됩니다. `continue` 문이 루프 내에서 실행되면, 반복문 루프 내의 나머지 코드들을 건너뛰고 다음 반복이 진행됩니다.

다음은 짝수를 출력하지 않도록 `continue` 문을 사용하는 `for` 반복문의 예입니다.

````python
for i in range(1, 10):
    if i % 2 == 0:  # 짝수면
        continue    # 건너뛰기
    print(i)
````
