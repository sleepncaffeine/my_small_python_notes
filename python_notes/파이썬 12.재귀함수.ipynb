# 재귀함수

<big><big>Recursion Function</big></big>

재귀 함수는 실행 중에 자신을 호출하는 함수입니다. 문제를 더 작은 하위 문제로 나누고 각 하위 문제는 동일한 함수를 재귀적으로 호출하여 해결하는 강력한 프로그래밍 기술입니다. 재귀 함수는 복잡한 문제를 더 간단한 자체 참조 단계로 나누어 해결하는 방법을 제공합니다.

## 재귀함수의 구조

재귀함수의 일반적인 구조는 다음과 같습니다:
```python
def recursive_function(parameters):
    if base_case:
        return base_case_value
    else:
        recursive_call = recursive_function(modified_parameters)
        return recursive_call
```
여기서 `parameters`는 함수의 입력되는 매개변수이고, `base_case`는 문제를 가장 단순한 형태 (기저 사례)로 축소되었는지 확인하는 조건입니다. `base_case_value`는 기저 사례의 해 입니다. 기저 사례를 만족하지 않으면 함수는 입력 매개변수를 수정하고 수정된 매개변수로 자기 자신을 재귀적으로 호출합니다.

다음은 재귀함수를 이용해서 팩토리얼($n!$)을 계산하는 함수입니다:

````python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

x = factorial(3)
print(x)
print(factorial(5))
print(factorial(7))
````

위 예시에서, 기저 사례는 `n==0`이고, 기저 사례의 값은 1입니다. 만약 기저 사례가 충족되지 않으면, 함수는 `n`과 `factorial(n-1)`의 재귀 호출 결과의 곱을 반환합니다.

예를 들어 $4!$을 계산하기 위해 함수는 다음과 같은 재귀 흐름을 가집니다.
```python
factorial(4) = 4 * factorial(3)
             = 4 * 3 * factorial(2)
             = 4 * 3 * 2 * factorial(1)
             = 4 * 3 * 2 * 1 * factorial(0)
             = 4 * 3 * 2 * 1 * 1
             = 24
```
함수는 4의 팩토리얼을 재귀적으로 3의 팩토리얼, 2의 팩토리얼, 1의 팩토리얼, 마지막으로 0의 팩토리얼을 계산하여 구합니다.

또한, 다음은 재귀함수의 이해를 돕기 위한 재밌는 예시입니다.<br>
출처: 백준 [17478 재귀함수가 뭔가요?](https://www.acmicpc.net/problem/17478)

````python
def recursive(m):
    print("_" * (4 * (n - m)) + '"재귀함수가 뭔가요?"')

    if not m:
        print("_" * (4 * (n - m)) + '"재귀함수는 자기 자신을 호출하는 함수라네"')
        print("_" * (4 * (n - m)) + "라고 답변하였지.")
        return

    print("_" * (4 * (n - m)) + '"잘 들어보게. 옛날옛날 한 산 꼭대기에 이세상 모든 지식을 통달한 선인이 있었어.')
    print("_" * (4 * (n - m)) + "마을 사람들은 모두 그 선인에게 수많은 질문을 했고, 모두 지혜롭게 대답해 주었지.")
    print("_" * (4 * (n - m)) + '그의 답은 대부분 옳았다고 하네. 그런데 어느 날, 그 선인에게 한 선비가 찾아와서 물었어."')
    recursive(m - 1)
    print("_" * (4 * (n - m)) + "라고 답변하였지.")


n = int(input())
print("어느 한 컴퓨터공학과 학생이 유명한 교수님을 찾아가 물었다.")
recursive(n)
````

## 재귀 vs 반복
재귀는 특정한 문제에 대해 효과적인 해결책이 될 수 있습니다. 그러나, 반복문을 사용한 해결책에 비하여 덜 효율적일 수 있습니다. 재귀 함수는 재귀 호출마다 새로운 스택 프레임을 생성하여 메모리를 소비하며, 과도한 재귀는 스택 오버플로 오류를 일으킬 수 있습니다. 반면에, 반복을 이용한 해결책은 반복문을 사용하여 문제를 반복적으로 처리하므로 메모리 측면에서 좀 더 효율적이고, 일부 문제에 대해서는 재귀보다 빠를 수 있습니다.

## 재귀함수의 주된 고려사항
재귀함수를 사용할 때에는 다음 요소들을 고려하는 것이 중요합니다.

* **기저 사례**: 재귀를 종료하기 위해 도달할 기저 사례가 있는지 확인합니다. 적절한 기본 사례가 없으면 재귀 함수가 무한정 실행되어 스택 오버플로 또는 과도한 메모리 사용이 발생할 수 있습니다.

* **재귀 호출**: 재귀 호출이 문제를 기저 사례에 더 가깝게 이동시키는 방식으로 입력을 수정하는지 확인합니다. 재귀 호출이 동일한 입력을 사용하거나 기저 사례로 수렴하지 않으면 무한 재귀로 이어집니다.<br>(`factorial(n)`을 계산하는데, `return n * factorial(n + 1)` 같은 짓을 하지 않기)

* 성능 고려: 재귀를 이용한 문제풀이는 깔쌈할 수 있지만 여러 함수 호출 및 스택 메모리 사용으로 인해 성능 오버헤드가 발생할 수 있습니다. 경우에 따라 걍 반복문을 쓰는 접근 방식이 더 효율적일 수 있습니다.

### 피보나치 수

````python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

result = fibonacci(6)
print(result) # 8
````

### 거듭제곱

````python
def power(base, exponent):
    if exponent == 0:
        return 1
    else:
        return base * power(base, exponent - 1)

result = power(2, 3)
print(result) # 8
````

### 이진탐색
이진탐색이란 오름차순으로 정렬된 정수의 리스트를 같은 크기의 두 부분 리스트로 나누고 필요한 부분에서만 탐색하도록 제한하여 원하는 원소를 찾는 알고리즘입니다. 리스트의 중간 부분에 찾는 원소가 있는지 확인하고, 없으면 위쪽에 있는지 아래쪽에 있는지 판단하여 맨 앞부터 검색하거나 중간부터 검색하는 방식으로, 엄청나게 빠른 탐색 알고리즘입니다.

````python
def binary_search(arr, target, low, high):
    if low > high:
        return -1 # 리스트에 없음
    else:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            return binary_search(arr, target, mid + 1, high)
        else:
            return binary_search(arr, target, low, mid - 1)

my_list = [1, 3, 5, 7, 9, 11, 13]
target = 7
result = binary_search(my_list, target, 0, len(my_list) - 1)
print(result) # 3 (목표값의 인덱스)
````

### [하노이의 탑](https://en.wikipedia.org/wiki/Tower_of_Hanoi)

>고대 인도 베나레스에 있는 한 사원의 이야기<br>
여기에는 다이아몬드로 이루어진 3개의 기둥이 있고, 그 기둥 중 하나에 가운데에 구멍이 나 기둥에 끼울 수 있게 된 64개의 크기가 각각 다른 황금 원반이 꽂혀 있다고 한다. 황금 원반은 가장 아래쪽에 있는 것이 가장 크고 위로 갈수록 점차 작아져 전체적으로 원추형의 탑을 이루고 있는데, 원반은 한 번에 하나씩만 옮길 수 있으며 작은 원반 위에 그보다 더 큰 원반을 옮길 수 없다.<br>
이 규칙으로 64개의 원반을 처음 놓여 있던 막대에서 다른 막대로 모두 옮기면 탑은 무너지고 세상의 종말이 온다고 한다.

재귀함수에 대해 공부할 때 예제로 쓰기 좋아 한번쯤은 보고 넘어간다.


````python
def tower_of_hanoi(n, source, auxiliary, destination):
    if n == 1:
        print(f"원반 1: {source} 에서 {destination} 로 옮긴다.")
    else:
        tower_of_hanoi(n - 1, source, destination, auxiliary)
        print(f"원반 {n}: {source} 에서 {destination} 로 옮긴다.")
        tower_of_hanoi(n - 1, auxiliary, source, destination)

tower_of_hanoi(3, 'A', 'B', 'C')
````

## [예제 3](https://www.acmicpc.net/problem/7113)

여러분이 할 일은 주어진 n과 m 값에 대해 윤민이가 위에서 설명한 방식으로 직사각형을 잘라서 얻은 사각형의 수를 계산하는 프로그램을 작성하는 것입니다.

아래는 그 프로그램입니다.<br>
[코드]
```python
def count_squares(n, m):
    if n == m:
```
