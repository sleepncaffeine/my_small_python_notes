# Python의 연산자

파이썬에서 연산자는 산술 연산, 비교 연산, 논리 연산 및 비트 연산과 같은 다양한 데이터 작업을 수행하는 데 사용됩니다.

## 산술 연산자

산술 연산자는 정수 및 부동([浮動](https://namu.wiki/w/%EC%BB%B4%ED%93%A8%ED%84%B0%EC%97%90%EC%84%9C%EC%9D%98%20%EC%88%98%20%ED%91%9C%ED%98%84#s-4)) 소수점 숫자와 같은 숫자 데이터 유형에서 수학 연산을 수행하는 데 사용됩니다.<br>Python에서 사용 가능한 다음 산술 연산자가 있습니다.
* `+` : 덧셈 연산자
* `-` : 뺄셈 연산자
* `*` : 곱셈 연산자
* `/` : 나눗셈 연산자
* `%` : 나머지 연산자
* `**` : 거듭제곱 연산자
* `//` : 버림 나눗셈 연산자 (얘는 나누고 나서 소수점 이하를 버립니다)

다음은 Python에서 산술 연산자를 사용하는 방법의 예시입니다.

```python
x = 10
y = 3

print(x + y)    # 13
print(x - y)    # 7
print(x * y)    # 30
print(x / y)    # 3.3333333333333335 (부동소수점 오차)
print(x % y)    # 1
print(x ** y)   # 1000
print(x // y)   # 3
```

## 비교 연산자

비교 연산자는 값들을 비교하고 비교 결과에 따라 불(부울, boolean, bool) 값(True 또는 False)을 반환하는 데 사용됩니다.<br>Python에서 다음과 같은 비교 연산자를 사용할 수 있습니다:
* `==` : 같음 연산자
* `!=` : 같지 않음 연산자
* `<` : 보다 작음 연산자
* `>` : 보다 큼 연산자
* `<=` : 작거나 같음 연산자
* `>=` : 크거나 같음 연산자

다음은 Python에서 비교 연산자를 사용하는 방법의 예시입니다.

```python
x = 10
y = 3

print(x == y)   # False
print(x != y)   # True
print(x < y)    # False
print(x > y)    # True
print(x <= y)   # False
print(x >= y)   # True
```

## 논리 연산자

논리 연산자는 불(bool) 값에 대해 논리 연산을 수행하는 데 사용됩니다.<br>Python에서 다음과 같은 논리 연산자가 사용됩니다:
* `and` : 논리 AND 연산자 (그리고)
* `or` : 논리 OR 연산자 (또는)
* `not` : 논리 NOT 연산자 (부정)

파이썬에서의 논리 연산자는 `||`라던가 `&&`처럼 생기지 않았습니다!

다음은 Python에서 논리 연산자를 사용하는 방법의 예시입니다.

```python
x = True
y = False

print(x and y)  # False
print(x or y)   # True
print(not x)    # False
```

## 비트 연산자 (볼필요 없음)

비트 연산자는 이진 값에 대해 연산을 수행하는 데 사용됩니다.<br>Python에서 다음과 같은 비트 연산자들이 사용 가능합니다.
* `&` : 비트 AND 연산자

|AND|0|1|
|:-:|:-:|:-:|
|0|0|0|
|1|0|1|

* `|` : 비트 OR 연산자

|OR|0|1|
|:-:|:-:|:-:|
|0|0|1|
|1|1|1|

* `^` : 비트 XOR 연산자

|XOR|0|1|
|:-:|:-:|:-:|
|0|0|1|
|1|1|0|

* `~` : 비트 NOT 연산자

|NOT|0|1|
|:-:|:-:|:-:|
||1||
|||0|

* `<<` : 비트 왼쪽 시프트 연산자

비트를 왼쪽으로 해당 수만큼 밈

* `>>` : 비트 오른쪽 시프트 연산자

비트를 오른쪽으로 해당 수만큼 밈

```python
x = 60    # 0011 1100
y = 13    # 0000 1101
print(x & y)   # 12 (0000 1100)
print(x | y)   # 61 (0011 1101)
print(x ^ y)   # 49 (0011 0001)
print(~x)      # -61 (1100 0011)
print(x << 2)  # 240 (1111 0000)
print(x >> 2)  # 15  (0000 1111)
```
