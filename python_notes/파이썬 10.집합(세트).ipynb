# 집합(세트)

집합(또는 세트)는 Python에서 **중복되지 않은** 요소들의 **순서 없는** 모음입니다. 또한, 이는 합집합, 교집합, 차이 등과 같은 집합의 연산을 수행하기 위한 효율적인 방법을 제공하는 자료형입니다.

## 집합 만들기

중괄호`{}` 안에 쉼표로 구분된 요소를 묶어서, 또는 `set()`을 사용해 집합을 만들 수 있습니다. 예를 들면 다음과 같습니다:

````python
# 숫자 집합 만들기
numbers = {1, 2, 3, 4, 5}
print(numbers)  # {1, 2, 3, 4, 5}

numbers_ = set([1, 2, 3, 4, 5])
print(numbers_)  # {1, 2, 3, 4, 5}

# 빈 집합 만들기
empty_set = set()
print(empty_set)  # set(), 딕셔너리와의 혼동을 고려해 빈 집합은 set()입니다.
````

첫 번째 예시에서는 요소 1, 2, 3, 4, 5를 포함하는 `numbers` 집합을 만듭니다. 두 번째 예시에서는 `set()` 함수를 사용하여 빈 집합을 만듭니다.

## 요소 넣다뺏다하기

Python의 집합은 변경 가능합니다. 즉, 요소를 동적으로 추가하고 제거할 수 있습니다.<br>
다음은 집합에서 요소를 추가하고 제거하기 위해 일반적으로 사용되는 몇 가지 방법입니다.

* `set.add(요소)`: 지정된 요소를 집합에 추가합니다.
* `set.update(요소들)`: 요소들을 집합에 추가합니다.
* `set.remove(요소)`: 집합에서 지정된 요소를 제거합니다. 요소가 존재하지 않으면 `KeyError`를 발생시킵니다.
* `set.discard(요소)`: 지정된 요소가 존재하는 경우 집합에서 제거합니다. 요소를 찾을 수 없는 경우 오류를 발생시키지 않습니다.
* `set.pop()`: 집합에서 임의의 요소를 제거하고 반환합니다. 세트가 비어 있으면 `KeyError`를 발생시킵니다.
* `set.clear()`: 집합에서 모든 요소를 ​​제거하여 비웁니다.

다음은 집합에서 요소를 추가하고 제거하는 방법을 보여주는 예입니다:

````python
fruits = {'apple', 'banana', 'orange'}
print(fruits)  # {'apple', 'banana', 'orange'}

# 새 과일 더하기
fruits.add('grape')
print(fruits)  # {'apple', 'banana', 'orange', 'grape'}

# 새 과일들 더하기
fruits.update(['cherry', 'strawberry'])  # {'apple', 'orange', 'banana', 'grape', 'cherry', 'strawberry'}
print(fruits)

# 바나나 지우기
fruits.remove('banana')
print(fruits)  # {'apple', 'orange', 'grape', 'cherry', 'strawberry'}

# discard 이용해 없는 과일 지우기
fruits.discard('mango')
print(fruits)  # {'apple', 'orange', 'grape', 'cherry', 'strawberry'}

# pop 이용해 아무 과일 하나 지우기
removed_fruit = fruits.pop()
print(removed_fruit)  # apple

# 집합 다 지우기
fruits.clear()
print(fruits)  # set()

# 결과는 런타임마다 달라질 수 있음
````

### 왜 위 코드 실행할때마다 특정 과일이 지워지고 집합 출력할때 순서는 어떻게 정해지는지 궁금하신 분들을 위한 설명

Python에서 집합은 순서가 지정되지 않은 자료형이므로 특정 순서가 없습니다. 그러나 CPython 인터프리터(가장 일반적으로 사용되는 Python 인터프리터)의 현재 구현은 집합에서 요소의 일관된 내부 순서를 따릅니다. 즉, `fruit.pop()`을 호출하면 어떤 식으로든 세트를 수정하지 않는 한 항상 동일한 요소를 제거하고 반환합니다.

집합의 요소 순서는 해시 값과 해시 테이블로 알려진 세트에서 사용하는 내부 데이터 구조를 기반으로 합니다. 해시 테이블은 집합에서 각 요소의 위치를 ​​결정하며 이 위치는 집합을 변경하지 않는 한 상대적으로 일관되게 유지됩니다.

이 동작은 Python의 현재 구현에만 해당되며 언어 사양에서 보장되지 않는다는 점에 유의해야 합니다. 다른 Python 구현 또는 Python의 향후 버전은 동일한 동작을 나타내지 않을 수 있습니다.

## 집합 연산

집합은 합집합, 교집합, 차이 및 대칭 차이와 같은 수학적 집합 연산을 수행할 수 있는 다양한 연산을 지원합니다.<br>
다음은 일반적으로 사용되는 몇 가지 집합 연산입니다.

* 합집합: `set1.union(set2)` 또는 `set1 | set2`
* 교집합: `set1.intersection(set2)` 또는 `set1 & set2`
* 차집합: `set1.difference(set2)` 또는 `set1 - set2`
* 대칭차집합: `set1.symmetric_difference(set2)` 또는 `set1 ^ set2`
* 하위 부분집합: `set1.issubset(set2)` 또는 `set1 <= set2`
* 상위 부분집합: `set1.issuperset(set2)` 또는 `set1 >= set2`

다음은 이것들을 사용하는 예시입니다:

````python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# 합집합
union_set = set1.union(set2)
print(union_set, set1 | set2)  # {1, 2, 3, 4, 5, 6, 7, 8}

# 교집합
intersection_set = set1.intersection(set2)
print(intersection_set, set1 & set2)  # {4, 5}

# 차집합
difference_set1 = set1.difference(set2)
difference_set2 = set2.difference(set1)
print(difference_set1, set1 - set2)  # {1, 2, 3}
print(difference_set2, set2 - set1)  # {6, 7, 8}

# 대칭차집합
symmetric_difference_set = set1.symmetric_difference(set2)
print(symmetric_difference_set, set1 ^ set2)  # {1, 2, 3, 6, 7, 8}

# 하위 부분집합
subset_check = set1.issubset(set2)
print(subset_check, set1 <= set2)  # False

# 상위 부분집합
superset_check = set1.issuperset(set2)
print(superset_check, set1 >= set2)  # False
````

## 집합의 일치 여부 확인

동일 연산자`==`과 `set1.isdisjoint(set2)`를 이용해 집합의 일치 여부를 확인할 수 있습니다.

`set1.isdisjoint(set2)`는 `set1`과 `set2`에 같은 요소가 없을 때 `True`를, 하나라도 있을 때 `False`를 반환합니다.

````python
s1 = set([1, 2, 3, 4, 5])
s2 = set([4, 5, 6, 7, 8])
s3 = {1, 2, 3, 4, 5}

if s1 == s2:
    print("s1과 s2는 같습니다.")
else:
    print("s1과 s2는 다릅니다.")

if s1 == s3:
    print("s1과 s3은 같습니다.")
else:
    print("s1과 s3은 다릅니다.")
````

````python
s1 = set([1, 2, 3, 4, 5])
s2 = set([4, 5, 6, 7, 8])
s3 = {1, 2, 3, 4, 5}
s4 = {4, 5, 6, 7, 8}

if s1.isdisjoint(s2):
    print("s1과 s2는 같은 요소가 하나도 없습니다.")
else:
    print("s1과 s2는 같은 요소가 적어도 하나는 있습니다.")

if s1.isdisjoint(s3):
    print("s1과 s3는 같은 요소가 하나도 없습니다.")
else:
    print("s1과 s3는 같은 요소가 적어도 하나는 있습니다.")

if s1.isdisjoint(s4):
    print("s1과 s4는 같은 요소가 하나도 없습니다.")
else:
    print("s1과 s4는 같은 요소가 적어도 하나는 있습니다.")
````

## 포함 관련

집합은 멤버십 테스트(포함하는지) 및 반복 기능을 제공합니다. 일반적으로 사용되는 몇 가지 기능은 다음과 같습니다.

* `element in set`: 지정된 요소가 집합에 있는지 확인합니다.
* `len(set)`: 집합의 요소 수를 반환합니다.
* `for element in set: ...`: 집합의 각 요소를 반복합니다.

다음은 예시입니다.

````python
fruits = {'apple', 'banana', 'orange'}

# 포함함?
print('apple' in fruits)  # True
print('grape' in fruits)  # False

# 길이
print(len(fruits))  # 3

# 반복
for fruit in fruits:
    print(fruit)  # apple, banana, orange
````

## 집합 내포

리스트 및 딕셔너리와 유사하게 세트는 간결하고 읽기 쉬운 구문을 사용하여 세트를 생성할 수 있는 집합 내포(컴프리헨션)도 지원합니다. 예를 들면 다음과 같습니다.

````python
# 집합 내포 이용해 제곱수의 집합 만들기
squares = {x ** 2 for x in range(1, 6)}
print(squares)  # {1, 4, 9, 16, 25}
````

위 예시에서는 집합 내포를 사용하여 1에서 5까지의 제곱한 수를 포함하는 `squares`라는 집합을 만듭니다.

````python
from io import StringIO
import sys

MASK1 = "" #@param {type:"string"}
MASK2 = "" #@param {type:"string"}
MASK3 = "" #@param {type:"string"}
MASK4 = "" #@param {type:"string"}
# set1 | set2
# set1 & set2
# set1 - set2
# set1 ^ set2

expected_output = """\
{1, 2, 4, 5, 7, 8, 9, 10, 11}
{1, 4, 9}
{8, 11, 7}
{2, 5, 7, 8, 10, 11}
"""

code = """
def set_operations(set1, set2):
    print( [MASK1] )
    print( [MASK2] )
    print( [MASK3] )
    print( [MASK4] )


A = {1, 4, 7, 8, 9, 11}
B = {1, 2, 4, 5, 9, 10}

set_operations(A, B)
""".replace("[MASK1]", MASK1).replace("[MASK2]", MASK2).replace("[MASK3]", MASK3).replace("[MASK4]", MASK4)

old_stdout = sys.stdout
sys.stdout = mystdout = StringIO()

error_flag = 0

try:
    exec(code)
except Exception as e:
    error_flag = 1

sys.stdout = old_stdout

result = mystdout.getvalue()

if result == expected_output:
    print("CORRECT")
else:
    print("WRONG")
    print(f"Expected output:\n{expected_output}")
    print(f"Your output:\n{result}")
    if error_flag == 1:
        print("ERROR OCCURRED")
````
