# Python의 `print()`로 출력하기

`print()` 메서드는 Python에서 출력을 표시하는 데 사용됩니다. 콘솔이나 기타 출력 장치에 텍스트나 데이터를 출력할 수 있는 내장 기능입니다. 이 섹션에서는 Python에서 `print()` 메서드를 사용하는 방법에 대한 기본 사항을 다룹니다.

## 기본 사용법

`print()`메서드의 기본 구문은 다음과 같습니다:
```python
print(object(s), sep=separator, end=end, file=file, flush=flush)
```

여기서 `object(s)`는 콘솔에 출력하려는 하나 이상의 객체를 나타냅니다. 쉼표로 구분하여 여러 객체를 전달할 수 있습니다. `sep` 매개 변수는 출력되는 객체들 사이의 구분자를 지정하는 데 사용됩니다(기본값은 공백 문자입니다). `end` 매개 변수는 출력 끝에 무엇을 출력해야 하는지를 지정하는 데 사용됩니다(기본값은 개행 문자입니다). `file` 매개 변수는 출력을 전송할 파일 객체를 지정하는 데 사용됩니다(기본값은 `sys.stdout`입니다). 마지막으로, `flush` 매개 변수는 출력을 즉시 플러시해야 하는지 여부를 지정하는 데 사용됩니다.

다음은 `print()` 메소드를 사용하여 간단한 문자열을 출력하는 방법의 예시입니다:

````python
print("Hello, World!")
````

`print()`에 변수를 넣으면 변수를 출력합니다.

````python
text = "Hello"
print(text)

number = 42
print(number)

print(text, number)
````

## `end`와 `sep` 매개 변수

위와 같이 `print()` 메서드는 주어진 내용을 출력한 뒤, 개행문자(`\n`)을 넣어 줄바꿈을 하고, 두 개 이상의 변수는 공백문자(` `)로 구분해 출력합니다.

이때, 개행문자 또는 공백 외에 다른 것을 넣고 싶다면, `end`와 `sep` 매개 변수를 사용할 수 있습니다.

### `end` 매개 변수

`end` 매개 변수는 출력 끝에 무엇을 출력해야 하는지를 지정하는 데 사용됩니다. 기본적으로 `end` 매개 변수는 개행 문자(`\n`)로 설정되어 있으며, 이는 각각의 `print()` 문이 새로운 줄로 끝남을 의미합니다. 그러나 `end` 매개 변수를 출력 끝에 출력하고자 하는 어떤 문자열이든 원하는대로 변경할 수 있습니다.

예를 들어, `end` 매개 변수를 사용하여 여러 `print()` 문을 한 줄에 출력할 수 있습니다:

````python
print("Hello", end=' ')
print("World")
````

`end` 매개 변수를 사용하여 + 연산자로 문자열을 연결할 수도 있습니다:

````python
name = "Alice"
age = 25
print("My name is " + name + " and I'm " + str(age) + " years old.", end=' ')
print("Nice to meet you!")
````

### `sep` 매개 변수

`sep` 매개 변수는 출력되는 객체들 간에 구분자(separator)를 지정하는 데 사용됩니다. 기본적으로 `sep` 매개 변수는 공백 문자(' ')로 설정되어 있으며, 이는 `print()` 메소드로 출력되는 각 객체들이 공백으로 구분되도록 합니다.

예를 들어:

````python
print("Alice", "Bob", "Charlie")
````

`sep` 매개 변수를 출력되는 객체들 간에 출력할 문자열로 변경할 수 있습니다. 예를 들어:

````python
print("Alice", "Bob", "Charlie", sep=', ')
````

## 출력 형식 지정

`print()` 메소드를 사용하면 여러 출력 형식 지정 옵션을 사용하여 출력을 서식화할 수 있습니다. 여기 일반적인 몇 가지 서식 지정 옵션이 있습니다:

### 1. 위치 인수를 사용한 서식 지정

문자열에서 중괄호 `{}`로 플레이스홀더(자리 표시자)를 사용하고, 값을 위치 인수를 사용하여 전달할 수 있습니다. 중괄호 안에는 인덱스나 변수 이름 등도 사용할 수 있습니다. 이 때, 값을 전달할 때는 `format()` 메소드를 사용합니다.

````python
name = "Alice"
age = 25
print("My name is {} and I'm {} years old.".format(name, age))
````

### 2. 키워드 인수를 사용한 서식 지정

이름이 지정된 키워드 인수를 사용하여도 플레이스홀더를 사용할 수 있습니다. 중괄호 `{}` 안에 키워드 인수의 이름을 넣고, `format()` 메소드에서 해당 인수의 값을 전달합니다. 이렇게 하면, 값을 전달할 때 인덱스를 명시할 필요 없이 키워드 인수의 이름으로 바로 전달할 수 있습니다.

````python
print("My name is {name} and I'm {age} years old.".format(name="Bob", age=30))
````

### 3. F-문자열을 사용한 서식 지정

Python 3.6부터는 f-문자열을 사용하여 문자열 내부에 표현식을 삽입하여 문자열 서식을 지정할 수 있습니다. f-문자열은 시작 따옴표 앞에 문자 'f' 또는 'F'가 있는 서식 지정된 문자열 리터럴입니다. 문자열 내부에서 중괄호 `{}`를 사용하여 런타임에 값을 치환할 표현식을 삽입할 수 있습니다.

````python
name = "Charlie"
age = 35
print(f"My name is {name} and I'm {age} years old.")
````

### 4. 숫자 서식화

숫자를 서식화하기 위해서, % 연산자와 함께 포맷 지정자를 사용할 수 있습니다. 다음은 몇 가지 예시입니다.

````python
x = 42
print("The answer is %d" % x)

y = 3.14159265358979
print("The value of π is approximately %.2f" % y)
````

첫번째 예시에서 `%d`는 `x`의 값을 정수로 포맷팅해야 함을 나타내는 포맷 지정자(format specifier)로 사용됩니다. 두번째 예시에서 `%.2f`는 `y`의 값을 소수점 이하 둘째자리까지 포맷팅해야 함을 나타내는 포맷 지정자입니다.



## 심화 사용법 (굳이 볼필요 없음)

### 1. 출력 리다이렉션

`print()` 메소드의 출력을 파일로 리다이렉션하려면 파일 객체를 `file` 매개변수로 전달하면 됩니다. 이렇게 하면 콘솔 대신 지정된 파일에 출력이 작성됩니다.

````python
with open("output.txt", "w") as f:
    print("Hello, World!", file=f)
````

### 2. 출력 버퍼링 비우기

기본적으로 `print()` 메소드는 출력을 버퍼링하고 새 줄 문자가 나타나거나 버퍼가 가득 찰 때만 콘솔에 기록합니다. 출력을 즉시 콘솔에 플러시하려면 `flush` 매개변수를 `True`로 설정할 수 있습니다. 이는 긴 시간이 걸리는 스크립트를 실행할 때와 같이 실시간으로 출력을 볼 필요가 있는 경우에 유용합니다.

````python
import time

for i in range(10):
    print(i, end=' ', flush=True)
    time.sleep(1)
````

### 3. 표준 오류 출력

출력 결과를 표준 출력뿐만 아니라 표준 오류 스트림에도 출력할 수 있습니다. 이를 위해서는 sys.stderr.write() 메서드를 사용합니다. 이를 사용하면 오류 메시지나 디버깅 정보 등을 출력할 수 있습니다.
```python
import sys

try:
    # 예외를 발생시킬 가능성이 있는 코드
except Exception as e:
    sys.stderr.write(f"An error occurred: {e}")
```
출력: `An error occurred: <예외 메세지>`

### 예제 1

첫 번째 줄에 `Hello, !DEA!`를 출력하는 코드를 작성하시오.

````python
import subprocess

입력 = "" #@param {type:"string"}

result = subprocess.run(['python', '-c', 입력], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
res = result.stdout.decode('utf-8').rstrip()
err = result.stderr.decode('utf-8').rstrip()
ans = "Hello, !DEA!"

if err:
    print("ERROR")
    print("해당 코드가 유효한 파이썬 코드인지, 그리고 예상한 출력이 나오는지 확인해주세요.")
    print("Error message:\n", err)
elif res == ans:
    print("CORRECT")
else:
    print("WRONG")
    print("Correct answer:\n", ans)
    print("Your answer:\n", res)
````

### 예제 2

코드의 실행 결과가 다음과 같도록 `[빈칸]`에 들어갈 알맞은 코드를 입력해 코드를 완성하시오.

실행 결과:<br>`Hello, Alice!`

코드:
```python
name = Alice
print(f'Hello, [빈칸]')
```

````python
def check_answer(user_input):
    expected_output = "Hello, Alice!"

    try:
        user_code = f"name = 'Alice'\noutput = f'Hello, {user_input}'\n"

        exec(user_code, globals())
        output = globals().get("output", "")

        if output.strip() == expected_output:
            return "CORRECT"
        else:
            return f"WRONG\nCorrect answer:\n'{expected_output}'\nYour output:\n'{output.strip()}'"

    except Exception as e:
        return f"""\
ERROR
해당 코드가 유효한 파이썬 코드인지, 그리고 예상한 출력이 나오는지 확인해주세요.
Error message:\n{e}\
"""

입력 = "" #@param {type:"string"}
print(check_answer(입력))
````
