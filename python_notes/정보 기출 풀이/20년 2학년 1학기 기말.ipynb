#### 정답

````python
class Car:
    def __init__(self, id, fuel):
        self.id = id
        self.fuel = fuel
    def drive(self):
        self.fuel -= 10
    def add_fuel(self, fuel):
        self.fuel += fuel
    def show_info(self):
        print("id : ", self.id)
        print("fuel : ", self.fuel)
class Truck(Car):
    def __init__(self, id, fuel, pack):
        super().__init__(id, fuel)
        self.pack = pack
    def add_pack(self, pack):
        self.pack += pack
    def show_info(self):
        super().show_info()
        print("pack : ", self.pack)
t = Truck("12광3456", 0, 0)
t.add_fuel(50)
t.add_pack(20)
t.drive()
t.show_info()
````

#### 정답

````python
import os, datetime, time, random

# 'log' 디렉토리의 존재 여부 확인
if not os.path.exists("log"):
    # 'log' 디렉토리 만들기
    os.mkdir("log")

# "count_log.txt"를 쓰기 모드로 생성
with open("log/count_log.txt", "w") as f:
    f.write("기록 시작\n")
    f.close()

# "count_log.txt"를 내용 추가 모드로 열기
with open("log/count_log.txt", "a") as f:
    for i in range(10):
        stamp = str(datetime.datetime.now())
        value = random.random()*1000000
        log_line = stamp + "\t" + str(value) + "\n"
        f.write(log_line)
````
