#### 정답

①: `s[j:j+i] == pre`<br>
②: `comp += str(cnt)+pre`<br>
③: `comp += pre`<br>
④: `pre = s[j:j+i]`<br>
⑤: `min(ans, len(comp))`

````python
def str_count(s):
    ans = len(s)
    for i in range(1, len(s)//2+1):
        comp=""
        pre = s[0:i]
        cnt = 1
        for j in range(i, len(s), i):
            # 이전 상태와 동일하면 횟수 증가
            if s[j:j+i] == pre:
                cnt += 1
            # 다른 문자열이 나오면 현재 문자열 저장 및 상태 초기화
            else:
                if cnt >= 2:
                    comp += str(cnt)+pre
                else:
                    comp += pre
                pre = s[j:j+i]
                cnt = 1
        # 남은 문자열 처리
        if cnt >= 2:
            comp += str(cnt)+pre
        else:
            comp += pre
        ans = min(ans, len(comp))
    return ans

print(str_count("aabbcc"))
print(str_count("abcabcabc"))
print(str_count("abcabcdede"))
print(str_count("abcabcabcabcdededede"))
print(str_count("xababcdcdababcdcd"))
````

### 10.다음은 숫자(0~9)로만 이루어진 문자열이 입력될 때, 왼쪽부터 오른쪽으로 하나씩 숫자를 확인하며 숫자 사이에 곱셈 연산자 또는 덧셈 연산자를 넣어 연산 결과가 가장 큰 수를 만들어 출력하는 프로그램이다. 실행 결과를 보고 빈 칸에 들어갈 내용을 완성하시오. [서술형 10점]

```python
input_num = input()
res = │     ① [2점]     │

for i in range(1, len(input_num)):
    num = │     ② [2점]     │
    if │        ③ [2점]        │:
```

[실행 결과]

출력 예시 2 - 19

#### 정답

①: `int(input_num[0])`<br>
②: `int(input_num[i])`<br>

````python
input_num = input()
res = int(input_num[0])

for i in range(1, len(input_num)):
    num = int(input_num[i])
    if num <= 1 or res <= 1:
        res += num
    else:
        res *= num

print(res)
````

### 11.람다식을 이용하여 [실행 결과]처럼 출력되도록 빈 칸에 들어갈 내용을 완성하시오. [단답형 1점, 서술형 6점]

```python
number_list = list(range(1, 21))

print("# 홀수 출력")
print(list(│ ① [단답형 1점] │(│      ② [서술형 2점]      │)))
print()

print("# 11 이상 17 이하")
print(list(│ ① [단답형 1점] │(│      ③ [서술형 2점]      │)))
print()

print("# 제곱 값이 70 미만만 출력")
print(list(│ ① [단답형 1점] │(│      ④ [서술형 2점]      │)))
print()
```
<결과>
```
# 홀수 출력
[1, 3, 5, 7, 9, 11, 13, 17, 19]

# 11 이상 17 이하 출력

# 제곱 값이 70 미만 출력
[1, 2, 3, 4, 5, 6, 7, 8]

```

#### 정답

①: `filter`<br>
②: `lambda x: x % 2 == 1, number_list`<br>
③: `lambda x: 11 <= x <= 17, number_list`<br>
④: `lambda x: x ** 2 < 70, number_list`

````python
number_list = list(range(1, 21))

print("# 홀수 출력")
print(list(filter(lambda x: x % 2 == 1, number_list)))
print()

print("# 11 이상 17 이하")
print(list(filter(lambda x: 11 <= x <= 17, number_list)))
print()

print("# 제곱 값이 70 미만만 출력")
print(list(filter(lambda x: x ** 2 < 70, number_list)))
print()
````
