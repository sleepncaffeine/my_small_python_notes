### 3.두 수 n, m을 입력받아 몫과 나머지를 리턴하는 함수 my_divmod 함수를 만드시오. [서술형 8점]

[조건]
* 함수 이름 my_divmod
* n, m > 0, n != m
* n, m 중 큰 수를 작은 수로 나눈다.
* 코드는 5줄 이내로 작성


#### 정답

````python
def my_divmod(n, m):
    div = n // m if n > m else m // n
    mod = n % m if n > m else m % n
    return (div, mod)

print("%d %d" % my_divmod(7, 2))
print("%d %d" % my_divmod(8, 15))
````

### 11.[코드 A]는 n 팩토리얼을 구하는 함수(fact_iter)이다. [코드 A]를 재귀 함수를 사용한 코드로 작성하시오. [서술형 6점]

[조건]
* 함수 이름 fact_recur
* 매개변수 n

[코드 A]
```python
def fact_iter(n):
    result = 1
    for i in range(1, n+1):
        result = result * i
    return result
```

[재귀 함수 활용 코드]
```python
# 코드 작성 부분
```

#### 정답

````python
def fact_recur(n):
    if n == 1 or n == 0:
        return 1
    return n * fact_recur(n-1)

print(fact_recur(8))
print(fact_recur(10))
````

① 오류의 명칭 (단답형):<br>
② 발생 원인 (서술형):

2) 1)에서 오류가 발생한 행 번호를 선택하고, 프로그램이 정상적으로 동작하도록 코드를 수정하시오. 단, counter 변수는 함수가 호출될 때마다 1씩 증가하는 변수임. [단답형 2점] ①, ② 각 1점

① 행 번호:<br>
② 수정 코드:

#### 정답

1)

① 오류의 명칭 (단답형): UnboundLocalError

② 발생 원인 (서술형): 함수 안에서 counter 변수를 먼저 참조하는 부분이 있기 때문에, 파이썬은 이 변수가 지역 변수로 정의되었다고 가정하고 해당 함수에서 이 변수를 참조하려고 시도한다. 하지만 counter 변수가 지역 변수에 정의되지 않았기 때문에, 해당 변수에 대한 할당이 실패한다. 따라서 UnboundLocalError가 발생한다.

2)

① 행 번호: 4

② 수정 코드:

````python
counter = 0

def fibo(n):
    global counter
    counter += 1
    if n == 1:
        return 1
    if n == 2:
        return 1
    else:
        return fibo(n-1) + fibo(n-2)

print(fibo(10))
print("함수호출 횟수 : ", counter)
````

만약 20이 입력되면 1부터 20까지의 소수는 2, 3, 5, 7, 11, 13, 17, 19로 총 8개입니다.

[입력 설명]

첫 줄에 자연수의 개수 N(2<=N<=200,000)이 주어진다.

[입력 예시]

20

[출력 예시]

8

[조건]

* 입력 변수 N으로 함.
* 15줄 이내로 작성함.
* 코드는 입력 부분, 처리 부분, 출력 부분으로 분리하여 작성함.

#### 정답

````python
# 입력 부분
N = int(input())
# 처리 부분
cnt = 0 # 소수의 개수
for i in range(2, N+1):
    is_prime = True # 소수 플래그
    for j in range(2, int(i**0.5)+1):
        if i % j == 0:
            is_prime = False
            break
    if is_prime:
        cnt += 1
# 출력 부분
print(cnt)
````
