##### 정답

````python
import random


def bingo_numbers_generator(n):
    bingo_numbers = list(range(1, n + 1))
    random.shuffle(bingo_numbers)
    for i in bingo_numbers:
        yield i


n = int(input("Input Bingo end number: "))
bingo_generator = bingo_numbers_generator(n)
for i in bingo_generator:
    print(i, end=" ")
````

#### 4.Define an iterator class named BingoNumberIterator that behaves like above. [12점]

```python

    print(i, end = ' ')
```

##### 정답

````python
import random


class BingoNumberIterator:
    def __init__(self, n):
        self.n = n
        self.numbers = list(range(1, n + 1))
        random.shuffle(self.numbers)
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index == self.n:
            raise StopIteration
        self.index += 1
        return self.numbers[self.index - 1]


n = int(input("Input Bingo end number: "))
bingo_iterator = BingoNumberIterator(n)
for i in bingo_iterator:
    print(i, end=" ")
````

#### 정답

````python
import time


def elapsed_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"Execution time: {execution_time:.10f} seconds")
        return result

    return wrapper


@elapsed_time
def sum_a_to_b(A, B):
    if A > B:
        A, B = B, A
    return sum(range(A, B + 1))


a, b = list(map(int, input("Enter two integers: ").split()))
print(f"The sum of {a} to {b} is {sum_a_to_b(a, b)}")
````

````python
class Date:
    months = [
        'Jan.', 'Feb.', 'Mar.', 'Apr.', 'May', 'Jun.',
        'Jul.', 'Aug.', 'Sep.', 'Oct.', 'Nov.', 'Dec.'
    ]

    def __init__(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day

    def isLeapYear(self):
        if (self.year % 4 == 0 and self.year % 100 != 0) or (self.year % 400 == 0):
            return True
        else:
            return False

    def __str__(self):
        return f"{Date.months[self.month - 1]} {self.day}, {self.year}"

    def __repr__(self):
        return f"Date({self.year}, {self.month}, {self.day})"

    @staticmethod
    def is_date_valid(date_str):
        year, month, day = map(int, date_str.split('-'))
        if (year > 0 and 1 <= month <= 12 and 1 <= day <= 31):
            return True
        else:
            return False

    @classmethod
    def create_date(cls, year, month, day):
        return cls(year, month, day)
````

* Roman numerals are usually writen largest to smallest from left to right. However, there are six cases in which the lft digit is substracted from the right digit.<br>
    <font face="monospace">
    * I can be placed before V(5) and X(10) to make IV(4) and IX(10).<br>

```python
if __name__ = = "__main__":
    Roman_numerals = ["III", "IV", "CXCV", "MCMLXXXVIII"]
    for i in Roman_numerals:
        print(f"{i] = {roman_to_int(i)}")
```

#### 정답

````python
def roman_to_int(roman):
    roman_dict = {"I": 1, "V": 5, "X": 10, "L": 50, "C": 100, "D": 500, "M": 1000}
    result = 0
    prev_value = 0

    for i in range(len(roman) - 1, -1, -1):
        current_value = roman_dict[roman[i]]
        if current_value < prev_value:
            result -= current_value
        else:
            result += current_value
        prev_value = current_value

    return result


if __name__ == "__main__":
    Roman_numerals = ["III", "IV", "VI", "CXCV", "MCMLXXXVIII"]
    for i in Roman_numerals:
        print(f"{i} = {roman_to_int(i)}")
````

#### 정답

````python
def check_parenthesis(formula):
    stack = []
    for ch in formula:
        if ch in ["(", "{", "["]:
            stack.append(ch)
        elif ch in [")", "}", "]"]:
            if len(stack) == 0:
                return False
            else:
                left = stack.pop()
                if (
                    (ch == ")" and left != "(")
                    or (ch == "}" and left != "{")
                    or (ch == "]" and left != "[")
                ):
                    return False
    return len(stack) == 0



if __name__ == "__main__":
    expressions = ["{(A+B)-3}+[{cos(x+y)+7}-1]*4",
                   "{(A+B)-3}+[{cos(x+y)+7}-1]*4)",
                   "{(A+B)-3}+[(cos(x+y)+7}-1]*4"]

    for exp in expressions:
        if check_parenthesis(exp):
            print(f"{exp} \n=> Parentheses in formulas are correct.")
        else:
            print(f"{exp} \n=> Parentheses in formulas are incorrect.")
````

````python
def find_unpaired_numbers(numbers):
    unpaired_numbers = []
    for i in numbers:
        if numbers.count(i) % 2 != 0:
            unpaired_numbers.append(i)
    return unpaired_numbers[0]


check_list = [[1, 2, 3, 2, 1, 1, 2, 1, 2],
              [8, 8, 6, 8, 8, 9, 8, 6, 8],
              [4, 7, 7, 4, 4, 4, 5, 4, 7, 7, 4],
              [6]]
for i in check_list:
    print(find_unpaired_numbers(i))
````
