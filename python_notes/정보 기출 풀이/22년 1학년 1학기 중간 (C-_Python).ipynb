### 10.다음은 전기세 계산 프로그램이다. [조건]을 고려하여 ㉠, ㉡에 알맞은 코드를 작성하시오. [12점]

[조건]
* 시간당 13,000원의 전기세가 소비되는 슈퍼컴퓨터가 있다. 4시간마다 누진세가 적용되어 9,000원이 추가된다. 40시간 이상 가동했을 때 누진세는 10,000원으로 오르고 전체 전기세의 10%를 더 지불하게 된다. 40시간 미만으로 가동했을 때는 2%의 전기세를 더 지불한다.
* 조건문 사용 시 `if`-`else`문을 사용한다.
* 실행예시는 다음과 같다.
```
─────────────                  ─────────────                  ─────────────
실행 예시 1                    실행 예시 2                    실행 예시 3
─────────────                  ─────────────                  ─────────────
가동 시간: 1                   가동 시간: 4                   가동 시간: 40
전기세: 13260원                전기세: 62220원                전기세: 682000원
```

```python
# 변수 지정
charge = 13000  # charge: 시간당 전기요금
time = 0 ; tax = 0 ; total = 0 # time: 시간, tax: 누진세, total: 전기세

㉠ 입력 안내 및 입력문 [2점]

㉡ 처리 및 출력 [10점]


```

#### 정답

````python
time = int(input("가동 시간: "))

if time < 40:
    tax = 9000
    total = time * charge + tax * (time // 4)
    total *= 1.02
else:
    tax = 10000
    total = time * charge + tax * (time // 4)
    total *= 1.1

print("전기세: %d원" % total)
````
