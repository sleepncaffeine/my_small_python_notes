### 5.다음[조건]을 고려하여 [실행결과]와 같이 출력되도록 ㉠~㉣에 알맞은 코드를 작성하시오. [9점]

[조건]
* 파스칼의 삼각형을 2차원 배열로 구현한 프로그램이다.
* 파스칼의 삼각형은 n번째 줄의 k번째 값을 $a_{nk}$라고 할 때,<br>
$a_{n1} = 1, a_{nn} = 1, a_{nk} = a_{(n-1)(k-1)} + a_{(n-1)k} (n, k ≥ 0)$으로 정의된다.

[실행결과]
```
   1
   1   1
   1   2   1
   1   3   3   1
   1   4   6   4   1
   1   5  10  10   5   1
   1   6  15  20  15   6   1
   1   7  21  35  35  21   7   1
   1   8  28  56  70  56  28   8   1
   1   9  36  84 126 126  84  36   9   1
```

```python
p = [[0 for i in range(0, 11)] for j in range(0, 11)]

for i in range(0, 11):
    for j in range(0, i):
        if |      ㉠      |:
            |      ㉡      |
        else:
            |      ㉢      |
        print("%4d" % |   ㉣   |)
    print()
```

#### 정답

㉠: `j == 0 or j == i`  
㉡: `p[i][j] = 1`  
㉢: `p[i][j] = p[i - 1][j - 1] + p[i - 1][j]`  
㉣: `p[i][j], end=""`

````python
p = [[0 for i in range(0, 11)] for j in range(0, 11)]

for i in range(0, 11):
    for j in range(0, i):
        if j == 0 or j == i:
            p[i][j] = 1
        else:
            p[i][j] = p[i - 1][j - 1] + p[i - 1][j]
        print("%4d" % p[i][j], end="")
    print()
````

### 6.다음은 이진탐색을 이용하여 데이터의 위치를 찾아주는 프로그램입니다. [조건]을 고려하여 알맞은 코드를 작성하시오. [13점]

[조건]
* 나머지 코드를 고려하여 탐색할 데이터의 배열 인덱스를 반환하는 함수를 정의할 것
* 이진탐색 결과를 반환하는 함수는 `binary_search()`이다.
* 탐색에 실패할 경우 `-1`을 반환해 출력한다.
* 실행결과가 [실행예제]와 같이 출력되도록 한다.

[실행예제]

|경우   |1 |2 |3 |4 |
|-------|--|--|--|--|
|입력 예|21|3|41|5|
|출력 예|5|0|8|-1|

```python
# binary_search() 함수 구현

SIZE = 10
grade = [3, 9, 11, 15, 19, 21, 23, 35, 41, 68]
key = int(input())

print(binary_search(grade, SIZE, key))
```

#### 정답

````python
def binary_search(grade, SIZE, key):
    low = 0
    high = SIZE - 1
    while low <= high:
        mid = (low + high) // 2
        if key == grade[mid]:
            return mid
        elif key < grade[mid]:
            high = mid - 1
        else:
            low = mid + 1
    return -1


SIZE = 10
grade = [3, 9, 11, 15, 19, 21, 23, 35, 41, 68]
keys = [21, 3, 41, 5]

for key in keys:
    print(binary_search(grade, SIZE, key))
````

```python
# ㉠ count_char() 함수 작성 [4점]

# ㉡ count_alphabet() 함수 작성 [6점]

# ㉢ count_word() 함수 작성 [7점]

arr = input()
print(
    "글자수: %d, 영문자수: %d, 단어수: %d"
    % (count_char(arr), count_alphabet(arr), count_word(arr))
)
```

#### 정답

````python
def count_char(arr):
    return len(arr)


def count_alphabet(arr):
    count = 0
    for i in arr:
        if i.isalpha():
            count += 1
    return count


def count_word(arr):
    count = 0
    for i in arr.split():
        for j in i:
            if j.isalpha():
                count += 1
                break
    return count


arr = input()
print(
    "글자수: %d, 영문자수: %d, 단어수: %d"
    % (count_char(arr), count_alphabet(arr), count_word(arr))
)
````

### 8.다음 [조건]을 고려하여 알맞은 코드를 작성하시오. [12점]

[조건]
* 광곽이는 동생을 위해 영어단어 받아쓰기를 간단하게 채점할 수 있는 프로그램을 만들었다.
* 2개의 단어를 입력받아 문자들을 비교해서 각 열의 문자가 같으면 O, 다르면 X를 출력할 것.
* 나머지 코드를 고려하여 문자를 비교해 O, X를 출력하는 `ox_check()` 함구를 구현할 것.
* [실행예시]와 같은 결과가 나오도록 작성하시오.

[실행예시]

|실행예시1|실행예시2|실행예시3|실행예시4|
|---------|---------|---------|---------|
|`Gwangju`<br>`gwangj`<br>`XOOOOX`|`science`<br>`csiense`<br>`XXOOOXO`|`programing`<br>`programming`<br>`OOOOOOOXXXX`|`phone`<br>`phone`<br>`OOOOO`|

```python
# ox_check() 함수 구현

s1 = input()
s2 = input()

ox_check(s1, s2)
```

#### 정답

````python
def ox_check(s1, s2):
    print(s1)
    print(s2)
    result = ""
    length = min(len(s1), len(s2))
    for i in range(length):
        if s1[i] == s2[i]:
            result += "O"
        else:
            result += "X"
    if len(s1) > len(s2):
        result += "X" * (len(s1) - len(s2))
    elif len(s2) > len(s1):
        result += "X" * (len(s2) - len(s1))
    print(result)


s1s = ["Gwangju", "science", "programing", "phone"]
s2s = ["gwangj", "csiense", "programming", "phone"]

for s1, s2 in zip(s1s, s2s):
    ox_check(s1, s2)
    print()
````

### 9.다음 [조건]을 고려하여 ㉠, ㉡에 알맞은 코드를 작성하시오. [18점]

[조건]
* 입력한 문자열과 왼쪽 화살표`<` or 오른족 화살표`>` 가 주어지면 각 방향으로 한 칸 이동한다.
* 예를 들어 문자열 `"ABCDE"`와 `>`를 입력 받으면 `EABCD`가 되고, `<<`를 입력 받으면 `CDEAB`가 된다.
* 문자열을 입력받고 `<`를 입력받으면 `left_func()`함수를 호출하여 문자열을 왼쪽으로 이동하고, `>`를 입력받으면 `right_func()`함수를 호출하여 문자열을 오른쪽으로 이동하는 프로그램을 구현할 것.
* [실행예시]와 같은 결과가 나오도록 작성하시오.

[실행예시]

|  |실행예시1|실행예시2|실행예시3|실행예시4|
|----|-------|--------|---------|----------|
|입력|`HAPPY`<br>`><<`|`printf`<br>`>>>`|`gwangju`<br>`>><`|`SCIENCE`<br>`>>>>>><`|
|출력|`APPYH`|`ntfpri`|`ugwangj`|`IENCESC`|

```python
# right_func() 함수 구현 [9점]

# left_func() 함수 구현 [9점]

a = input()
b = input()

for i in range(len(b)):
    if b[i] == '<':
        left_func(a, len(a))
    elif b[i] == '>':
        right_func(a, len(a))

print(a)
```

#### 정답

````python
# 문제의 의도 (c스럽게)

def right_func(s, n):
    s_list = list(s)
    temp = s_list[-1]
    for i in range(n - 1, 0, -1):
        s_list[i] = s_list[i - 1]
    s_list[0] = temp
    return "".join(s_list)


def left_func(s, n):
    s_list = list(s)
    temp = s_list[0]
    for i in range(0, n - 1):
        s_list[i] = s_list[i + 1]
    s_list[-1] = temp
    return "".join(s_list)


a_list = ["HAPPY", "printf", "gwangju", "SCIENCE"]
b_list = ["><<", ">>>", ">><", ">>>>>><"]

for a, b in zip(a_list, b_list):
    for i in range(len(b)):
        if b[i] == "<":
            a = left_func(a, len(a))
        elif b[i] == ">":
            a = right_func(a, len(a))

    print(a)
````

````python
# 파이썬스럽게

def right_func(s, n):
    s = s[-1] + s[:-1]
    return s

def left_func(s, n):
    s = s[1:] + s[0]
    return s

a_list = ["HAPPY", "printf", "gwangju", "SCIENCE"]
b_list = ["><<", ">>>", ">><", ">>>>>><"]

for a, b in zip(a_list, b_list):
    for i in range(len(b)):
        if b[i] == "<":
            a = left_func(a, len(a))
        elif b[i] == ">":
            a = right_func(a, len(a))

    print(a)
````
