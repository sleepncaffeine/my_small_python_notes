# 변수의 정의 및 할당

Python에서 변수는 값을 가리키는 이름입니다. 할당 연산자 (`=`)를 사용하여 변수를 정의하고 값을 할당할 수 있습니다.

다음은 하나의 예시입니다:

````python
x = 10
````

이 예제에서는 변수 `x`를 정의하고 값 `10`을 할당하고 있습니다.

또한 다른 변수에 할당된 값을 다른 변수에 할당할 수도 있습니다:

````python
x = 10

y = x
````

이 예시에서는 변수 `x`와 `y`를 정의하고 있습니다. 변수 `x`에는 값 `10`을 할당하고, 그 뒤에 `x`의 값을 `y`에 할당하고 있습니다.

# 변수의 이름짓기

Python에서 변수를 이름을 정할 때는 몇 가지 규칙을 따르는 것이 좋습니다:

* 변수 이름은 설명적이고 의미 있게 지어야 합니다. 변수의 목적을 명확하게 나타내는 이름을 사용하십시오.

* 변수 이름은 소문자로 시작하는 것이 좋습니다. 이름이 여러 단어로 구성된 경우 밑줄을 사용하여 구분하십시오 (예: `my_variable`).

* 예약어(또는 키워드, `if`, `else`, `while` 등)를 변수 이름으로 사용하지 마십시오.

* 상수에 대해서는 대문자를 사용하여 구분하기 쉽게 만드십시오. (예: `PI`, `GRAVITY` 등).

다음은 적절한 변수 이름의 몇 가지 예시입니다:
```python
my_variable = 10
my_other_variable = "Hello, world!"
my_constant = 3.14159
```
그리고 다음은 부적절한 변수 이름의 예시들입니다:
```python
3d_model = "Invalid variable name" # 변수의 이름은 숫자로 시작할 수 없습니다.
if = 10   # "if"는 예약어입니다.
```
파이썬 코드에서 변수 이름을 지을 때 규칙을 따르는 것이 중요합니다. 이렇게 하면 코드를 읽고 이해하기 쉬워지며, 코드를 함께 작업하는 다른 사람들에게도 도움이 됩니다.

추가로, 아래 코드를 통해 *keyword*, 즉 Python의 예약어로 변수의 이름으로써 쓰이지 못하는 단어를 알 수 있습니다.

````python
import keyword

kwlist = keyword.kwlist
for i in range(0,len(kwlist)):
    print("[{:14}]".format(kwlist[i]), end=" ")
    if (i+1)%5==0: print()
````

# Python에서의 상수

파이썬에서 상수(Constant)는 프로그램 전체에서 값이 변하지 않는 변수를 의미합니다. 상수는 일반적으로 수학상수나 물리상수와 같이 알려진 값이나 고정된 값을 저장하는 데 사용됩니다.

일부 다른 프로그래밍 언어와 달리, 파이썬은 내장된 상수 개념이 없습니다. 하지만 파이썬에서는 상수로 사용될 변수의 이름을 대문자로 작성하는 것이 관례이며, 이는 해당 값이 변경되지 않아야 함을 나타내는 방법입니다.

다음은 예시입니다:
```python
PI = 3.14159
GRAVITY = 9.8
```
이 예제에서는 `PI`와 `GRAVITY` 두 개의 변수를 정의하고 있습니다. 이러한 변수들은 상수로 사용될 것이기 때문에 대문자로 이름을 지정했습니다.

그러나 변수 이름에 대문자를 사용하는 것은 실제로 변수의 값을 변경하지 못하게 하는 것은 아닙니다. 그러나 Python 커뮤니티에서 널리 인정되는 관례이며, 코드를 더 읽기 쉽고 이해하기 쉽게 만드는 데 도움이 됩니다.

Python 코드에서 상수를 정의할 때는 상수의 목적을 정확하게 반영하고 의미 있는 이름을 선택하는 것이 중요합니다. 또한, 상수의 목적과 관련된 핵심 정보와 상수의 값을 설명하는 코드 주석을 포함하는 것이 좋은 습관입니다.
