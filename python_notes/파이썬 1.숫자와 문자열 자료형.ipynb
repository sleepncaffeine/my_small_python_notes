# Python의 자료형(데이터 유형, 데이터 타입)

프로그래밍에서 자료형은 프로그램에서 사용되는 데이터의 유형 또는 종류를 나타냅니다. 파이썬은 동적으로 유형이 지정된 언어이며, 변수를 선언할 때 데이터 유형을 지정할 필요가 없습니다. 대신, 파이썬은 변수에 할당하는 값에 기반하여 적절한 자료형형을 자동으로 지정합니다.

## 1. 수치/숫자 자료형

숫자 자료형은 파이썬에서 숫자를 나타내는 데 사용됩니다. 파이썬의 세 가지 기본 숫자 자료형은 다음과 같습니다:

* 정수 (`int`): 정수는 소수점 없이 양수와 음수의 정수입니다. 예: 2, 4, -6, 0.

* 부동 소수점 숫자 (`float`): 부동 소수점 숫자는 소수점을 가지는 실수입니다. 예: 3.14, -2.5, 0.0.

* 복소수 (`complex`): 복소수는 `a + bj` 형태로 나타내며, 여기서 `a`와 `b`는 모두 부동 소수점 숫자 또는 정수이고, `j`는 허수를 나타냅니다. 예: 2+3j, -4-2j.

## 2. 텍스트 자료형

텍스트 자료형은 문자열을 나타내는 데 사용됩니다. Python에서 텍스트 자료형은 문자를 단일 또는 이중 따옴표로 감싸는 방식으로 표현되며, 문자열(string, str)이라고 합니다. 예를 들어: "Hello, World!", 'This is a string'.

또한, `len()`함수는 문자열의 길이를 반환하는 함수입니다.<br>
<small>`len()`함수는 리스트에서도 같은 방식으로 사용할 수 있습니다.</small>

````python
print(len("Hello"))
````

## 3. 불 자료형

파이썬에서 불(부울, Boolean, Bool) 자료형은 True와 False 값을 나타내는 데 사용됩니다. 불 값은 종종 조건문에서 사용되며, 특정 조건이 참(True)인 경우에만 코드 블록이 실행됩니다.

### 문자열의 인덱싱과 슬라이싱

파이썬에서 문자열은 문자의 순서 있는 시퀀스입니다. 이는 문자열의 각 문자가 숫자로 할당되는 것을 의미합니다. 첫 번째 문자에 대해서는 0, 두 번째 문자에 대해서는 1로 시작하여 계속해서 번호가 할당됩니다. 이 번호를 사용하여 문자열의 개별 문자에 접근할 수 있습니다.

#### 인덱싱

문자열에서 특정 문자에 접근하려면 인덱싱을 사용할 수 있습니다. 인덱싱은 대괄호 `[]`를 사용하며, 대괄호 안에 인덱스 번호를 넣습니다. 다음은 예시입니다:

````python
s = "Hello, World!"
print(s[0])
````

이 예시에서는 우리가 인덱스 `0`을 사용하여 문자열 `s`의 첫 번째 문자에 접근하고 있습니다.

문자열의 뒤에서부터 문자에 접근하려면, 마지막 문자에 대해 `-1`부터 시작하는 음수 인덱싱을 사용할 수 있습니다. 다음은 예시입니다:

````python
s = "Hello, World!"
print(s[-1])
````

이 예제에서는 인덱스 `-1`을 사용하여 문자열 s의 마지막 문자에 접근합니다.

#### 슬라이싱

슬라이싱은 문자열의 일부분을 추출하는 방법입니다. 콜론 `:`으로 구분된 인덱스 범위를 사용하여 문자열을 자를 수 있습니다. 예시를 보겠습니다:

````python
s = "Hello, World!"
print(s[0:5])
````

이 예제에서는 문자열 `s`를 인덱스 `0`부터 인덱스 `4`까지 슬라이싱하여 `"Hello"`라는 부분 문자열을 얻습니다. 슬라이싱에서 마지막 인덱스는 포함되지 않습니다. 따라서 인덱스 `5`의 문자는 슬라이싱된 부분 문자열에 포함되지 않습니다. (인덱스 `0`이상 `5`미만으로 이해하면 쉽습니다.)

또한, 위에서의 인덱싱과 같이 문자열의 뒤에서부터 슬라이싱하려면 음수 인덱싱을 사용할 수 있습니다. 다음은 예시입니다.

````python
s = "Hello, World!"
print(s[-6:-1])
````

이 예제에서는, 문자열 `s`를 뒤에서부터 여섯 번째 문자부터 두 번째 문자까지 자르면 `"World"`라는 부분 문자열이 생성됩니다. 마찬가지로 시작 인덱스는 포함되고 끝 인덱스는 제외됩니다.

마지막으로, 인덱스 중 하나 이상을 생략하여 문자열을 시작 또는 끝에서부터 자를 수 있습니다. 다음은 예시입니다:

````python
s = "Hello, World!"
print(s[:5])
print(s[7:])
````

첫 예시에서는, 문자열 `s`를 처음부터 `4`번 인덱스까지 자르면서 `"Hello"` 부분 문자열을 얻습니다. 두 번째에서는, 문자열 `s`를 `7`번 인덱스부터 끝까지 자르면서 `"World!"` 부분 문자열을 얻습니다.

아, 마지막이 아니였군요.

파이썬에서 문자열을 슬라이싱할 때는 `[start:end]` 구문을 사용하여 추출할 문자의 범위를 지정할 수 있습니다. 이 구문은 슬라이스의 시작 인덱스(포함)와 끝 인덱스(제외)를 지정합니다.

`[start:end:step]` 구문을 사용하여 간격 값을 지정할 수도 있습니다. 간격 값은 말 그대로 추출될 문자 사이의 간격을 결정합니다. 예시를 살펴보겠습니다:

````python
s = "0123456789"
print(s[2:8:2])
````

이 예제에서는 인덱스 `2`에서 시작하고 인덱스 `7`에서 끝나는 (마지막 인덱스는 제외되는) 문자열 `s`를 슬라이싱하며, 간격 값을 `2`로 설정합니다. 이는 인덱스 `2`와 인덱스 `7` 사이에서 문자를 두 칸에 한 번씩 추출하므로 `"246"`이라는 부분 문자열이 반환됩니다.

간격 값을 입력하지 않으면, 기본값인 `1`로 설정됩니다.

문자열에서 문자를 역순으로 추출하기 위해 음수의 간격 값을 사용할 수 있습니다. 다음은 예시입니다:

````python
s = "0123456789"
print(s[8:2:-2])
````

이 예시에서는 문자열 `s`를 인덱스 `8`에서 시작하여 인덱스 `3`에서 끝나는 음수 간격 값이 `-2`인 슬라이싱을 수행합니다. 이는 우리가 인덱스 `8`과 인덱스 `3` 사이에서 역순으로 매 두 번째 문자를 추출하고 이를 연결하여 부분 문자열 `'864'`을 얻는 것을 의미합니다.

음수 간격 값을 사용할 때 시작과 끝 인덱스를 올바르게 지정해야만 원하는 출력을 얻을 수 있음에 유의하십시오. 예를 들어:

````python
s = "0123456789"
print(s[2:8:-2])
# 출력: ""
````

이 예시에서는 문자열 `s`를 인덱스 `2`에서 시작하여 인덱스 `7`에서 끝나는 음수 간격 값이 `-2`인 역순 슬라이싱을 시도합니다. 그러나 시작 인덱스가 끝 인덱스보다 크기 때문에 슬라이스는 비어 있고 출력은 빈 문자열 `""`입니다.

### 문자열 메소드

파이썬에는 문자열을 막 갖고놀 수 있는 문자열을 위한 메소드들이 있습니다. 다음은 많이 쓰이는 문자열 메소들입니다.

* `len()`: `len()`함수는 문자열의 길이를 리턴합니다.

````python
my_string = "Hello World"
print(len(my_string))  # 11
````

* `split()`: `split()`메소드는 지정된 구분자를 기준으로 문자열을 여러 개의 하위 문자열로 분할하는 데 사용됩니다(리스트로 리턴합니다). 기본적으로 구분자는 공백입니다.

````python
my_string = "Hello World"
print(my_string.split())  # ['Hello', 'World']

# 공백 말고 다른 문자 기준으로 분할하기기
my_string = "apple,banana,orange"
print(my_string.split(","))  # ['apple', 'banana', 'orange']
````

* `in`: `in` 키워드는 ~~<font size=2em>메소드는 아니긴 한데 걍 여따 넣을게요</font>~~ 문자열 내에 부분 문자열이 있는지 여부를 확인하는 데 사용됩니다. 이는 불 값(True 또는 False)을 반환합니다.

````python
my_string = "Hello World"
print("World" in my_string)  # True
print("world" in my_string)  # False
````

* `replace()`: `replace()` 메소드는 문자열에서 하나 이상의 문자열을 새로운 문자열로 교체하는 데 사용됩니다.

````python
my_string = "Hello World"
new_string = my_string.replace("World", "Python")
print(new_string)  # Hello Python

# 두 개 이상도 한번에 교체됩니다
my_string = "Hello Hello World"
new_string = my_string.replace("Hello", "Bye")
print(new_string) # Bye Bye world
````

* `upper()`: `upper()` 메소드는 문자열의 모든 문자를 대문자로 변환하는 데 사용됩니다.

````python
my_string = "Hello World"
print(my_string.upper())  # HELLO WORLD
````

* `lower()`: `lower()` 메소드는 문자열의 모든 문자를 소문자료 변환하는 데 사용됩니다.

````python
my_string = "Hello World"
print(my_string.lower())  # hello world
````

* `strip()`: `strip()` 메소드는 문자열의 시작과 끝에서 공백 문자를 제거하는 데 사용됩니다.

````python
my_string = "  Hello World  "
print('"' + my_string + '"')          # "  Hello World  "
print('"' + my_string.strip() + '"')  # "Hello World"
````
