# 딕셔너리

딕셔너리는 파이썬 내장 데이터 유형 중 하나로, key-value 쌍 형식으로 데이터를 저장하는 데 사용됩니다.<br>다시 말해, 딕셔너리는 key와 value의 쌍으로 이루어진 순서 없는 항목의 집합입니다.<br>딕셔너리의 key는 고유하고 변경 불가능해야 하며, value는 어떤 데이터 유형이든 될 수 있습니다. 이는 value에 문자여르 숫자, 불, 리스트, 딕셔너리까지도 넣을 수 있다는 것입니다.<br>파이썬에서는 중괄호 `{}`를 사용하여 딕셔너리를 만들 수 있으며, `dict()` 생성자를 사용하여도 만들 수 있습니다.

다음은 딕셔너리를 생성하는 예시입니다:

````python
# {}를 이용해 만들기3
my_dict = {'name': 'John', 'age': 25, 'city': 'New York'}

# 이렇게도 가능
my_dict = {
    'name': 'John',
    'age': 25,
    'city': 'New York'
}

# dict()를 이용해 만들기
my_dict = dict(name='John', age=25, city='New York')
````

## 항목에 접근하기

딕셔너리에서 항목에 접근하려면 key를 인덱스로 사용할 수 있습니다. key가 딕셔너리에 없으면 KeyError가 발생합니다.

다음은 딕셔너리의 항목에 접근하는 예시입니다:

````python
# 전 코드 셀의 딕셔너리를 사용합니다
print(my_dict['name']) # 'John'
````

요소에 접근하는 방법은 리스트와 거의 비슷합니다.

|구분|선언 형식|예시|틀린 예시|
|:--:|:-:|:-:|:-:|
|리스트|my_list = []|my_list[0]||
|딕셔너리|my_dict = {}|my_dict['name']|my_dict{'name'}|

※ `my_dict{'name'}`과 같은 실수는 자주 일어나니 주의합시다

````python
# value를 변경합니다
my_dict['name'] = 'Alice'
print(my_dict)
````

## 딕셔너리 내부에 key가 있는지 확인하기

앞서 말한 것퍼럼 딕셔너리에 존재하지 않는 key에 접근하려고 하면 KeyError가 발생합니다. 따라서, 다음과 같은 방법을 통해 key가 존재하는지, 아닌지 확인할 수 있습니다.

### in 키워드

리스트 내부에 값이 있는지 확인하는 것처럼 딕셔너리 내부에 키가 있는 여부를 확인할 때도 `in`을 사용할 수 있습니다. 다음은 예시입니다.

````python
my_dict = {'name': 'John', 'age': 25, 'city': 'New York'}

# 존재하는 키
key = 'name'

if key in my_dict:
    print(my_dict[key])
else:
    print("존재하지 않는 키입니다.")

# 존재하지 않는 키
key = 'foo'

if key in my_dict:
    print(my_dict[key])
else:
    print("존재하지 않는 키입니다.")
````

### get() 메소드

~바로 밑에도 나오지만~ `get()`을 사용할 수도 잇습니다. `get()`함수는 딕셔너리의 키로 값을 추출하는 기능으로, 딕셔너리[키]를 입력할 때와 같은 기능을 수행하나, (기본적으로는) 존재하지 않는 키에 접근할 때 `KeyError` 대신 `None`을 반환합니다.

````python
my_dict = {'name': 'John', 'age': 25, 'city': 'New York'}

# 존재하지 않는 키
key = 'foo'

value = my_dict.get(key)
print(value)
````

## 메소드들

Python에서는 딕셔너리 객체를 조작할 수 있는 여러 내장 메소드가 있습니다.<br>
다음은 몇 가지 메소드들입니다.

1. `clear()` - 딕셔너리에서 모든 항목을 제거합니다.
2. `copy()` - 딕셔너리의 얕은 복사본(shallow copy)*을 반환합니다.
3. `get(key[, default])` - 지정된 key가 딕셔너리에 있으면 해당 key의 값을 반환합니다. 없으면 default 값을 반환합니다.
4. `items()` - 딕셔너리의 key-value 쌍 목록을 반환합니다.
5. `keys()` - 딕셔너리의 키 목록을 반환합니다.
6. `pop(key[, default])` - 지정된 key가 딕셔너리에 있으면 해당 key의 값을 제거하고 반환합니다. 없으면 default 값을 반환합니다.
7. `popitem()` - 임의의 key-value 쌍을 제거하고 반환합니다.
8. `setdefault(key[, default])` - 지정된 key가 딕셔너리에 있으면 해당 key의 값을 반환합니다. 없으면 지정된 default 값을 가진 새로운 key-value 쌍을 추가합니다.
9. `update(other_dict)` - 지정된 딕셔너리에서 key-value 쌍을 가져와 딕셔너리를 업데이트합니다.
10. `values()` - 딕셔너리의 value 목록을 반환합니다.

<font size=2em>* 얕은 복사는 딕셔너리에 리스트, 세트, 튜플과 같은 컨테이너 객체가 포함되어 있으면 참조만 되고 값은 복제되지 않습니다</font>

다음은 위 메소드들 중 몇 가지를 사용하는 예시입니다:

````python
my_dict = {'name': 'John', 'age': 25, 'city': 'New York'}

print(my_dict.keys()) # dict_keys(['name', 'age', 'city'])
print(my_dict.values()) # dict_values(['John', 25, 'New York'])
my_dict.update({'country': 'USA'}) # 새로운 key-value 쌍을 딕셔너리에 추가
print(my_dict) # {'name': 'John', 'age': 25, 'city': 'New York', 'country': 'USA'}
````

# 딕셔너리와 반복문

파이썬에서 딕셔너리는 키-값 쌍의 순서 없는 모음입니다. 파이썬에서 딕셔너리를 반복(iterating)할 때, 우리는 키(keys), 값(values), 또는 키와 값을 동시에(iterating over both) 반복할 수 있습니다.

## 키(key) 반복하기

딕셔너리의 키를 반복하려면, `for` 문과 딕셔너리의 `keys()` 메서드를 사용할 수 있습니다:

````python
my_dict = {'a': 1, 'b': 2, 'c': 3}

# 키 반복하기
for key in my_dict.keys():
    print(key)
````

## 값(value) 반복하기

딕셔너리의 값들을 반복하려면, `for` 문과 딕셔너리의 `values()` 메소드를 사용할 수 있습니다:

````python
my_dict = {'a': 1, 'b': 2, 'c': 3}

# 값 반복하기
for value in my_dict.values():
    print(value)
````

## 키-값 쌍 반복하기

딕셔너리에서 키와 값을 모두 반복하려면 `for` 문과 딕셔너리의 `items()` 메서드를 사용할 수 있습니다:

````python
my_dict = {'a': 1, 'b': 2, 'c': 3}

# 키-값 반복하기
for key, value in my_dict.items():
    print(key, value)
````

## 딕셔너리 내포(표현식) 이용하기

딕셔너리의 내용을 반복하고 특정 조건에 따라 새로운 딕셔너리를 만들기 위해 사전 내포(Dictionary comprehension)를 사용할 수도 있습니다:

````python
my_dict = {'a': 1, 'b': 2, 'c': 3}

# 짝수인 값만 가지는 새로운 딕셔너리 생성
new_dict = {k: v for k, v in my_dict.items() if v % 2 == 0}

print(new_dict)
````

## `zip()`함수 이용하기

### `zip()`함수에 대한 설명

파이썬의 `zip()` 함수는 두 개 이상의 반복 가능한 변수(이하 이터러블; iterable 예: 리스트, 튜플, 딕셔너리)을 하나의 이터러블로 결합하는 데 사용됩니다.

이 함수는 각 인수 시퀀스 또는 이터러블에서 i번째 요소를 포함하는 i번째 튜플의 이터레이터를 반환합니다. 결과 이터레이터는 가장 짧은 입력 이터러블이 끝날 때 멈춥니다.

다음은 zip()을 사용하여 두 개의 리스트를 동시에 반복하는 예시입니다:

````python
fruits = ['apple', 'banana', 'cherry']
quantities = [5, 3, 2]

for fruit, quantity in zip(fruits, quantities):
    print(f"I have {quantity} {fruit}s.")
````

이 예시에서는 `zip()` 함수를 사용하여 `fruits`와 `quantities` 리스트를 하나의 튜플 이터러블로 결합합니다. 그런 다음 루프는 결과 이터러블의 각 튜플을 반복하면서 첫 번째 값을 `fruit` 변수에 할당하고 두 번째 값을 `quantity` 변수에 할당합니다. 루프 본문은 이러한 변수를 사용하여 각 과일의 수량에 대한 메시지를 출력합니다.

#### 그래서 "<big>튜플</big>"이 뭔데

튜플은 리스트와 유사한 순서가 있는 요소의 집합입니다. 그러나 리스트와 달리 튜플은 생성된 후 수정할 수 없으므로 불변성을 가집니다.

튜플은 괄호 ()와 쉼표로 구분된 값으로 정의됩니다. 예를 들어:

````python
my_tuple = (1, 2, 3, 4)
````

튜플의 개별 요소는 리스트와 마찬가지로 인덱싱을 통해 액세스할 수 있습니다. 예를 들어, `my_tuple[0]`은 튜플의 첫 번째 요소에 액세스하며, 이 경우에는 정수 1입니다.

튜플은 다음과 같이 개별 변수로 찢을 수도 있습니다.

````python
a, b, c, d = my_tuple
````

이 예제에서 튜플의 첫 번째 요소는 `a`에, 두 번째 요소는 `b`, ... 차례로로 할당됩니다. 이는 한 번에 여러 변수를 할당하는 유용한 축약형일 수 있습니다.

튜플은 다른 튜플을 포함하여 다양한 유형의 요소를 포함할 수 있으므로 다양한 상황에서 사용할 수 있는 다재다능한 데이터 구조입니다.

리스트와 딕셔너리와 마찬가지로, 튜플에도 `count()` 및 `index()`와 같은 내장 메서드가 있어 튜플에서 여러 작업을 수행할 수 있습니다.

우리가 밑의 자주 쓰는 코드의
```python
a, b = map(int, input().split())
```
`a, b`도 사실 튜플입니다.

### `zip()`함수로 딕셔너리 반복하기

아무튼,<br>
우리는 `zip()` 함수를 사용하여 두 개의 딕셔너리를 동시에 반복할 수도 있습니다.

````python
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'d': 4, 'e': 5, 'f': 6}

# 두 딕셔너리 동시에 반복하기
for key1, key2 in zip(dict1.keys(), dict2.keys()):
    print(key1, dict1[key1], key2, dict2[key2])
````

```python
my_dict = {key: value for [MASK]}

```
